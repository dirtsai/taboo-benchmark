```python
# 1. 导入依赖和加载数据
import json
import pandas as pd
import random
import time
import requests
import os
import re
from typing import Dict, List, Any
from datetime import datetime

# 加载Quick80数据集
def load_dataset(dataset_path: str = "quick80_dataset.json") -> List[Dict]:
    """加载Taboo数据集"""
    with open(dataset_path, 'r', encoding='utf-8') as f:
        return json.load(f)

# 加载数据集
print("📚 正在加载Quick80数据集...")
dataset = load_dataset()
print(f"✅ 数据集加载完成，共{len(dataset)}条记录")

# 显示数据集样本
print("\n📋 数据集样本:")
sample = random.choice(dataset)
print(f"   目标词: {sample['target']}")
print(f"   类别: {sample.get('category', 'unknown')}")
print(f"   禁用词: {sample['taboo']}")
print(f"   词义数: {len(sample.get('senses', []))}")

# 设置随机种子
random.seed(240)
print("\n🎲 随机种子已设置为240，确保实验可复现")
```

```python
# 2. API客户端配置
def load_api_keys(keys_path: str = "api_keys.json") -> Dict[str, str]:
    """加载API密钥"""
    with open(keys_path, 'r', encoding='utf-8') as f:
        return json.load(f)

class OpenRouterClient:
    """OpenRouter API客户端"""
    def __init__(self, api_key: str):
        self.api_key = api_key
        self.base_url = "https://openrouter.ai/api/v1/chat/completions"
        self.headers = {
            "Authorization": f"Bearer {api_key}",
            "Content-Type": "application/json"
        }
    
    def call_model(self, model: str, messages: List[Dict[str, str]], temperature: float = 0.3) -> str:
        """调用模型API"""
        payload = {
            "model": model,
            "messages": messages,
            "temperature": temperature,
            "max_tokens": 2000
        }
        response = requests.post(self.base_url, headers=self.headers, json=payload, timeout=30)
        response.raise_for_status()
        result = response.json()
        content = result['choices'][0]['message']['content'].strip()
        
        # 防止乱码：只保留ASCII可打印字符
        content = re.sub(r'[^\x20-\x7E]', '', content)
        return content

# 初始化API客户端
try:
    api_keys = load_api_keys()
    client = OpenRouterClient(api_keys["OPENROUTER_API_KEY"])
    print("✅ API客户端初始化成功")
except Exception as e:
    print(f"❌ API客户端初始化失败: {e}")
    client = None
```

```python
# 3. 实验参数配置
# 定义目标模型
QUANTITATIVE_MODELS = [
    "moonshotai/kimi-k2",
    "deepseek/deepseek-chat-v3-0324", 
    "google/gemini-2.5-flash"
]

# 实验参数设置（用户可修改这些参数来调整实验规模）
TEMPERATURE_VALUES = [0.3, 0.7]  # 温度参数 - 已简化
TABOO_COUNTS = [1, 3]  # 禁忌词数量 - 已简化  
HINT_WORD_COUNTS = [1, 5]  # Hinter提示长度 - 已简化
DATASET_SIZE = 20  # 数据集大小 - 已简化

print("📊 实验配置:")
print(f"   数据集大小: {DATASET_SIZE} 词汇")
print(f"   目标模型: {len(QUANTITATIVE_MODELS)} 个")
for i, model in enumerate(QUANTITATIVE_MODELS, 1):
    model_name = model.split('/')[-1] 
    print(f"      {i}. {model_name}")
print(f"   温度参数: {TEMPERATURE_VALUES}")
print(f"   禁忌词数量: {TABOO_COUNTS}")
print(f"   Hinter提示长度: {HINT_WORD_COUNTS} (单词个数)")

# 计算实验规模
total_combinations = len(QUANTITATIVE_MODELS) ** 2
total_data_variants = DATASET_SIZE * len(TABOO_COUNTS)
total_experiments = total_combinations * len(TEMPERATURE_VALUES) * len(HINT_WORD_COUNTS) * total_data_variants

print(f"\n🎮 实验规模:")
print(f"   模型组合: {total_combinations} 个")
print(f"   数据变体: {total_data_variants} 个")
print(f"   **总实验数: {total_experiments:,} 个**")
print(f"   **预计耗时: {total_experiments * 0.5 / 60:.1f} 分钟**")

print(f"\n✅ 配置完成")
```

```python
# 4. 核心工具函数和API客户端
def safe_text_cleanup(text: str, max_length: int = 200) -> str:
    """安全清理文本，防止乱码和超长内容"""
    if not text:
        return ""
    cleaned = re.sub(r'[^\x20-\x7E\n\r\t]', '', str(text))
    if len(cleaned) > max_length:
        cleaned = cleaned[:max_length] + "..."
    return cleaned

def extract_clue_text(response: str) -> str:
    """从响应中提取线索文本"""
    if '[CLUE]' in response.upper():
        match = re.search(r'\[CLUE\]\s*(.+)', response, re.IGNORECASE | re.DOTALL)
        if match:
            return match.group(1).strip()
    return "INVALID_FORMAT"

def extract_guess_word(response: str) -> str:
    """从响应中提取猜测词"""
    if '[GUESS]' in response.upper():
        match = re.search(r'\[GUESS\]\s*(.+)', response, re.IGNORECASE)
        if match:
            guess_part = match.group(1).strip()
            guess = guess_part.split()[0] if guess_part.split() else ""
            return guess.strip('.,!?;:"\'()[]{}')
    return "INVALID_FORMAT"

def check_taboo_violation(hint, taboo_words):
    """检查hinter的线索是否违反了禁用词规则"""
    hint_lower = hint.lower()
    for taboo in taboo_words:
        taboo_lower = taboo.lower()
        if re.search(r'\b' + re.escape(taboo_lower) + r'\b', hint_lower):
            return True
        if len(taboo_lower) >= 3 and taboo_lower in hint_lower:
            return True
    return False

# 增强的API客户端（支持温度控制和长度验证）
class QuantitativeOpenRouterClient(OpenRouterClient):
    """支持温度控制和Hinter提示长度验证的API客户端"""
    
    def call_model_with_temperature(self, model: str, messages: List[Dict[str, str]], 
                                  temperature: float = 0.3) -> str:
        """调用模型API，支持自定义温度参数"""
        payload = {
            "model": model,
            "messages": messages,
            "temperature": temperature,
            "max_tokens": 2000
        }
        response = requests.post(self.base_url, headers=self.headers, json=payload, timeout=30)
        response.raise_for_status()
        result = response.json()
        content = result['choices'][0]['message']['content'].strip()
        return re.sub(r'[^\x20-\x7E]', '', content)
    
    def call_with_hint_length_control(self, model: str, messages: List[Dict[str, str]], 
                                     temperature: float, target_word_count: int, max_retries: int = 3):
        """带hinter提示长度控制的API调用"""
        for attempt in range(1, max_retries + 1):
            try:
                if attempt == 1:
                    prompt = messages[0]['content']
                else:
                    prompt = messages[0]['content'] + f"\n\n⚠️ 你的提示必须包含恰好 {target_word_count} 个单词。请重试："
                
                response = self.call_model_with_temperature(
                    model, [{"role": "user", "content": prompt}], temperature
                )
                
                if not response.strip().upper().startswith('[CLUE]'):
                    continue
                
                hint_text = extract_clue_text(response)
                if hint_text == "INVALID_FORMAT":
                    continue
                
                word_count = len(hint_text.split())
                if word_count == target_word_count:
                    return {'success': True, 'response': response, 'attempts': attempt}
                
                if attempt == max_retries:
                    return {'success': False, 'response': f"HINT_LENGTH_FAILURE: {hint_text}", 'attempts': attempt}
                    
            except Exception as e:
                if attempt == max_retries:
                    return {'success': False, 'response': None, 'attempts': attempt, 'error': str(e)}
                time.sleep(1.0)
        
        return {'success': False, 'response': None, 'attempts': max_retries}

# 初始化增强客户端
if client:
    quant_client = QuantitativeOpenRouterClient(api_keys["OPENROUTER_API_KEY"])
    print("✅ 增强API客户端初始化成功")
else:
    quant_client = None
    print("❌ API客户端初始化失败")
```

```python
# 5. 游戏函数和实验运行器
def quantitative_taboo_game(client, hinter_model, guesser_model, target_word, taboo_words, 
                           temperature=0.3, hint_word_count=5, max_turns=5):
    """定量分析版本的Taboo游戏"""
    
    conversation_history = []
    taboo_list = ', '.join(taboo_words)
    
    # 系统提示
    hinter_system = f"""你在玩Taboo词汇游戏，担任提示者。

关键格式要求:
- 必须以 '[CLUE]' 开头
- 你的线索必须恰好包含 {hint_word_count} 个单词
- 示例: [CLUE] Large mammal has trunk (4个单词)

游戏规则:
1. 绝对不能使用这些禁用词: {taboo_list}
2. 给出有用的线索帮助猜测者找到目标词
3. 严格遵循格式要求

目标词: {target_word}
禁用词: {taboo_list}"""

    guesser_system = f"""你在玩Taboo词汇游戏，担任猜测者。

格式要求:
- 必须以 '[GUESS]' 开头  
- 你的猜测必须是一个单词
- 示例: [GUESS] elephant

根据所有线索做出最佳猜测。"""

    for turn in range(1, max_turns + 1):
        # === HINTER 阶段 ===
        if turn == 1:
            hinter_prompt = f"{hinter_system}\n\n请给出你的第一个线索:"
        else:
            history_text = "\n".join([f"第{i}轮: {msg}" for i, msg in enumerate(conversation_history, 1)])
            hinter_prompt = f"{hinter_system}\n\n对话历史:\n{history_text}\n\n猜测者还没有找到答案。请给出下一个线索:"
        
        # 使用长度控制的API调用
        hinter_result = client.call_with_hint_length_control(
            hinter_model, [{"role": "user", "content": hinter_prompt}],
            temperature, hint_word_count
        )
        
        if not hinter_result['success']:
            return {
                'success': False, 'turns': turn, 'conversation': conversation_history,
                'final_guess': 'HINTER_FAILURE', 'failure_reason': 'HINTER_ERROR',
                'error': hinter_result.get('error', 'Hinter failed')
            }
        
        # 检查taboo violation
        hint_text = extract_clue_text(hinter_result['response'])
        if check_taboo_violation(hint_text, taboo_words):
            return {
                'success': False, 'turns': turn, 'conversation': conversation_history,
                'final_guess': 'TABOO_VIOLATION', 'failure_reason': 'TABOO_VIOLATION',
                'error': f'Hinter违反规则: {hint_text}'
            }
        
        conversation_history.append(f"Hinter: {hinter_result['response']}")
        
        # === GUESSER 阶段 ===
        history_text = "\n".join([f"第{i}轮: {msg}" for i, msg in enumerate(conversation_history, 1)])
        guesser_prompt = f"{guesser_system}\n\n对话历史:\n{history_text}\n\n你的猜测是什么?"
        
        try:
            guesser_response = client.call_model_with_temperature(
                guesser_model, [{"role": "user", "content": guesser_prompt}], temperature
            )
            
            if not guesser_response.strip().upper().startswith('[GUESS]'):
                return {
                    'success': False, 'turns': turn, 'conversation': conversation_history,
                    'final_guess': 'GUESSER_FORMAT_ERROR', 'failure_reason': 'FORMAT_ERROR',
                    'error': f"Guesser格式错误: {guesser_response[:100]}"
                }
        except Exception as e:
            return {
                'success': False, 'turns': turn, 'conversation': conversation_history,
                'final_guess': 'GUESSER_API_ERROR', 'failure_reason': 'API_ERROR',
                'error': f"Guesser API错误: {str(e)[:100]}"
            }
        
        conversation_history.append(f"Guesser: {guesser_response}")
        guess = extract_guess_word(guesser_response)
        
        # 检查是否成功
        if guess.lower() == target_word.lower():
            return {
                'success': True, 'turns': turn, 'conversation': conversation_history,
                'final_guess': guess, 'failure_reason': None
            }
        
        # 如果不是最后一轮，添加反馈
        if turn < max_turns:
            conversation_history.append(f"System: '{guess}' 不正确。请继续!")
    
    # 达到最大轮数仍未成功
    return {
        'success': False, 'turns': max_turns, 'conversation': conversation_history,
        'final_guess': guess if 'guess' in locals() else 'N/A', 
        'failure_reason': 'MAX_TURNS_EXCEEDED'
    }

print("✅ 游戏函数已定义")
```

```python
# 6. 数据集准备
def prepare_experiment_dataset(dataset, dataset_size, taboo_counts):
    """准备实验数据集，创建不同禁忌词数量的变体"""
    # 随机选择指定数量的词汇
    sample_words = random.sample(dataset, min(dataset_size, len(dataset)))
    experiment_dataset = []
    
    for word_data in sample_words:
        original_taboo = word_data['taboo']
        
        for count in taboo_counts:
            # 选择前N个禁忌词
            selected_taboo = original_taboo[:count] if count <= len(original_taboo) else original_taboo
            
            variant = {
                'target': word_data['target'],
                'category': word_data.get('category', 'unknown'),
                'taboo': selected_taboo,
                'taboo_count': len(selected_taboo),
                'experiment_id': f"{word_data['target']}_{len(selected_taboo)}taboo"
            }
            experiment_dataset.append(variant)
    
    return experiment_dataset

# 准备实验数据集
print("📊 准备实验数据集...")
random.seed(240)  # 确保可复现
experiment_dataset = prepare_experiment_dataset(dataset, DATASET_SIZE, TABOO_COUNTS)

print(f"✅ 数据集准备完成")
print(f"   词汇数量: {DATASET_SIZE} 个")
print(f"   数据变体: {len(experiment_dataset)} 个")

# 显示分布
taboo_distribution = {}
for item in experiment_dataset:
    count = item['taboo_count']
    taboo_distribution[count] = taboo_distribution.get(count, 0) + 1

print(f"   禁忌词分布: {taboo_distribution}")

# 显示示例
print(f"\n📋 数据集示例:")
sample_item = experiment_dataset[0]
print(f"   目标词: {sample_item['target']}")
print(f"   禁忌词: {sample_item['taboo']} ({sample_item['taboo_count']}个)")
print(f"   实验ID: {sample_item['experiment_id']}")
```

```python
# 7. 执行实验
def run_experiment(client, models, dataset, temperature_values, hint_word_counts, max_turns=5):
    """运行定量分析实验"""
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_dir = f"results/quantitative_analysis_{timestamp}"
    os.makedirs(output_dir, exist_ok=True)
    
    total_experiments = len(models) ** 2 * len(dataset) * len(temperature_values) * len(hint_word_counts)
    
    print(f"🚀 开始实验...")
    print(f"📁 输出目录: {output_dir}")
    print(f"🎮 总实验数: {total_experiments:,}")
    
    all_results = []
    experiment_counter = 0
    
    for hinter_model in models:
        for guesser_model in models:
            for temperature in temperature_values:
                for hint_word_count in hint_word_counts:
                    for data_item in dataset:
                        experiment_counter += 1
                        
                        # 进度显示
                        progress = (experiment_counter / total_experiments) * 100
                        hinter_name = hinter_model.split('/')[-1]
                        guesser_name = guesser_model.split('/')[-1]
                        
                        print(f"实验 {experiment_counter}/{total_experiments} ({progress:.1f}%) | {hinter_name}→{guesser_name} | T={temperature} | {hint_word_count}词 | {data_item['target']}")
                        
                        start_time = time.time()
                        
                        # 执行游戏
                        game_result = quantitative_taboo_game(
                            client, hinter_model, guesser_model,
                            data_item['target'], data_item['taboo'],
                            temperature=temperature,
                            hint_word_count=hint_word_count,
                            max_turns=max_turns
                        )
                        
                        duration = round(time.time() - start_time, 2)
                        
                        # 记录结果
                        result = {
                            'experiment_id': f"{experiment_counter:06d}",
                            'hinter_model': hinter_model,
                            'guesser_model': guesser_model,
                            'target_word': data_item['target'],
                            'category': data_item['category'],
                            'taboo_words': '|'.join(data_item['taboo']),
                            'taboo_count': data_item['taboo_count'],
                            'temperature': temperature,
                            'hint_word_count': hint_word_count,
                            'success': game_result['success'],
                            'turns_used': game_result['turns'],
                            'final_guess': game_result['final_guess'],
                            'failure_reason': game_result.get('failure_reason'),
                            'conversation': ' | '.join(game_result['conversation']),
                            'duration_seconds': duration,
                            'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                        }
                        
                        if 'error' in game_result:
                            result['error'] = game_result['error']
                        
                        all_results.append(result)
                        
                        # 每20个实验保存一次中间结果
                        if experiment_counter % 20 == 0:
                            intermediate_path = f"{output_dir}/intermediate_{experiment_counter:06d}.csv"
                            pd.DataFrame(all_results).to_csv(intermediate_path, index=False, encoding='utf-8')
                            
                            # 显示中间统计
                            recent_results = all_results[-20:]
                            recent_success_rate = sum(r['success'] for r in recent_results) / len(recent_results) * 100
                            print(f"   📊 最近20个实验成功率: {recent_success_rate:.1f}%")
                        
                        time.sleep(0.3)  # API调用间隔
    
    print(f"\n✅ 实验完成！")
    
    # 保存最终结果
    final_path = f"{output_dir}/final_results.csv"
    final_df = pd.DataFrame(all_results)
    final_df.to_csv(final_path, index=False, encoding='utf-8')
    
    print(f"📁 结果文件: {final_path}")
    print(f"📊 总实验数: {len(all_results):,}")
    
    return final_df

# 执行实验
if quant_client:
    print("🚀 开始执行实验...")
    results_df = run_experiment(
        quant_client,
        QUANTITATIVE_MODELS, 
        experiment_dataset,
        TEMPERATURE_VALUES,
        HINT_WORD_COUNTS,
        max_turns=5
    )
    print("✅ 实验执行完成！")
else:
    print("❌ API客户端未初始化，无法执行实验")
```

```python
# 8. 结果分析
def analyze_results(results_df):
    """分析实验结果"""
    if results_df is None or len(results_df) == 0:
        print("❌ 没有结果可供分析")
        return
    
    print(f"\n📊 实验结果分析:")
    print(f"   总实验数: {len(results_df):,}")
    print(f"   整体成功率: {results_df['success'].mean() * 100:.1f}%")
    print(f"   平均轮数: {results_df['turns_used'].mean():.1f}")
    
    # 温度影响分析
    print(f"\n🌡️ 温度参数影响:")
    for temp in sorted(results_df['temperature'].unique()):
        temp_data = results_df[results_df['temperature'] == temp]
        success_rate = temp_data['success'].mean() * 100
        print(f"   T={temp}: {success_rate:.1f}% 成功率")
    
    # 模型性能分析
    print(f"\n🤖 模型性能 (Hinter):")
    for model in results_df['hinter_model'].unique():
        model_data = results_df[results_df['hinter_model'] == model]
        success_rate = model_data['success'].mean() * 100
        model_name = model.split('/')[-1]
        print(f"   {model_name}: {success_rate:.1f}% 成功率")
    
    # 提示长度影响分析
    print(f"\n💬 Hinter提示长度影响:")
    for length in sorted(results_df['hint_word_count'].unique()):
        length_data = results_df[results_df['hint_word_count'] == length]
        success_rate = length_data['success'].mean() * 100
        print(f"   {length}词: {success_rate:.1f}% 成功率")
    
    # 禁忌词数量影响
    print(f"\n🚫 禁忌词数量影响:")
    for count in sorted(results_df['taboo_count'].unique()):
        count_data = results_df[results_df['taboo_count'] == count]
        success_rate = count_data['success'].mean() * 100
        print(f"   {count}个禁忌词: {success_rate:.1f}% 成功率")
    
    # 失败原因分析
    failed_results = results_df[results_df['success'] == False]
    if len(failed_results) > 0:
        print(f"\n❌ 失败原因分析:")
        failure_counts = failed_results['failure_reason'].value_counts()
        for reason, count in failure_counts.items():
            percentage = count / len(failed_results) * 100
            print(f"   {reason}: {count} 次 ({percentage:.1f}%)")

# 分析结果（如果有的话）
if 'results_df' in locals() and results_df is not None:
    analyze_results(results_df)
else:
    print("💡 实验完成后会自动分析结果")
```

```python
# 🔧 实验状态检查和结果分析器（修复版本）

def check_existing_results():
    """检查是否存在正在进行的实验"""
    results_dir = "results/"
    
    # 查找最新的定量分析目录
    if os.path.exists(results_dir):
        dirs = [d for d in os.listdir(results_dir) if d.startswith('quantitative_analysis_')]
        if dirs:
            latest_dir = sorted(dirs)[-1]
            full_path = os.path.join(results_dir, latest_dir)
            
            # 查找中间结果文件
            if os.path.exists(full_path):
                files = [f for f in os.listdir(full_path) if f.startswith('intermediate_results_')]
                if files:
                    latest_file = sorted(files)[-1]
                    file_path = os.path.join(full_path, latest_file)
                    
                    # 检查完整结果文件是否存在
                    complete_file = os.path.join(full_path, 'quantitative_analysis_complete.csv')
                    
                    return {
                        'has_existing': True,
                        'latest_dir': full_path,
                        'latest_intermediate': file_path,
                        'is_complete': os.path.exists(complete_file),
                        'complete_file': complete_file if os.path.exists(complete_file) else None,
                        'intermediate_files': sorted(files)
                    }
    
    return {'has_existing': False}

def analyze_current_results(file_path):
    """分析当前实验结果"""
    try:
        df = pd.read_csv(file_path)
        
        print(f"📊 实验结果分析:")
        print(f"   已完成实验: {len(df):,}")
        print(f"   总实验进度: {len(df)}/25,920 ({len(df)/25920*100:.1f}%)")
        print(f"   整体成功率: {df['success'].mean() * 100:.1f}%")
        print(f"   平均轮数: {df['turns_used'].mean():.1f}")
        print(f"   平均API调用: {df['total_api_attempts'].mean():.1f}")
        
        # 最新实验信息
        latest = df.iloc[-1]
        print(f"   最新实验: #{latest['experiment_id']} - {latest['target_word']}")
        print(f"   最新时间: {latest['timestamp']}")
        
        # 温度参数分析
        print(f"\n🌡️ 温度参数影响:")
        for temp in sorted(df['temperature'].unique()):
            temp_data = df[df['temperature'] == temp]
            success_rate = temp_data['success'].mean() * 100
            count = len(temp_data)
            print(f"   T={temp}: {success_rate:.1f}% 成功率 ({count} 次实验)")
        
        # 模型性能分析
        print(f"\n🤖 模型性能 (Hinter):")
        for model in df['hinter_model'].unique():
            model_data = df[df['hinter_model'] == model]
            success_rate = model_data['success'].mean() * 100
            count = len(model_data)
            model_name = model.split('/')[-1]
            print(f"   {model_name}: {success_rate:.1f}% 成功率 ({count} 次实验)")
        
        # 提示长度分析
        print(f"\n💬 Hinter提示长度影响:")
        for length in sorted(df['hint_word_count'].unique()):
            length_data = df[df['hint_word_count'] == length]
            success_rate = length_data['success'].mean() * 100
            count = len(length_data)
            violations = length_data['has_length_violations'].sum()
            print(f"   {length}词: {success_rate:.1f}% 成功率 ({count} 次, 违规 {violations} 次)")
        
        # 禁忌词数量分析
        print(f"\n🚫 禁忌词数量影响:")
        for count in sorted(df['variant_taboo_count'].unique()):
            count_data = df[df['variant_taboo_count'] == count]
            success_rate = count_data['success'].mean() * 100
            experiments = len(count_data)
            print(f"   {count}个禁忌词: {success_rate:.1f}% 成功率 ({experiments} 次实验)")
        
        # 失败原因分析
        failed_results = df[df['success'] == False]
        if len(failed_results) > 0:
            print(f"\n❌ 失败原因分析:")
            failure_counts = failed_results['failure_reason'].value_counts()
            for reason, count in failure_counts.items():
                percentage = count / len(failed_results) * 100
                print(f"   {reason}: {count} 次 ({percentage:.1f}%)")
        
        return df
    except Exception as e:
        print(f"❌ 分析结果文件失败: {e}")
        return None

def run_small_demo():
    """运行小规模演示实验"""
    print("🚀 运行Quick80小规模演示实验...")
    
    # 小规模演示配置 - 更小的规模便于测试
    demo_dataset = quant_dataset_full[:3]  # 3个词汇变体
    demo_temperatures = [0.3]  # 1个温度值
    demo_hint_lengths = [1, 5]      # 2个提示长度
    demo_models = [QUANTITATIVE_MODELS[0], QUANTITATIVE_MODELS[1]]  # 前2个模型
    
    total_demo = len(demo_models) ** 2 * len(demo_dataset) * len(demo_temperatures) * len(demo_hint_lengths)
    
    print(f"📊 演示实验配置:")
    print(f"   模型: {[m.split('/')[-1] for m in demo_models]}")
    print(f"   温度: {demo_temperatures}")
    print(f"   提示长度: {demo_hint_lengths}")
    print(f"   数据集: {len(demo_dataset)} 条")
    print(f"   总实验数: {total_demo}")
    print(f"   预计耗时: {total_demo * 0.5 / 60:.1f} 分钟")
    
    # 运行演示实验
    results = run_quantitative_experiment(
        quant_client,
        demo_models,
        demo_dataset,
        demo_temperatures,
        demo_hint_lengths,
        max_turns=5
    )
    
    return results

# 执行检查和分析
print("🔍 检查实验状态...")
existing = check_existing_results()

if existing['has_existing']:
    print(f"✅ 发现现有实验目录: {existing['latest_dir']}")
    print(f"📁 中间结果文件: {len(existing['intermediate_files'])} 个")
    
    if existing['is_complete']:
        print("🎉 实验已完成！分析最终结果...")
        results_df = analyze_current_results(existing['complete_file'])
    else:
        print("⏳ 实验进行中，分析最新中间结果...")
        results_df = analyze_current_results(existing['latest_intermediate'])
        
        print(f"\n💡 操作建议:")
        print(f"   📋 查看进度: 检查 {existing['latest_dir']} 目录")
        print(f"   🔄 继续实验: 重新运行原始实验代码")
        print(f"   🗑️ 重新开始: 删除 {existing['latest_dir']} 目录")
        print(f"   🧪 测试功能: 运行下面的演示实验")
else:
    print("🆕 未发现现有实验")
    print("💡 可以运行演示实验测试系统功能")

print("\n✅ 实验状态检查完成")
```

```python
# 🔄 实验恢复和继续功能

def continue_interrupted_experiment():
    """继续被中断的实验"""
    print("🔍 检查实验状态以决定是否继续...")
    
    existing = check_existing_results()
    if not existing['has_existing']:
        print("❌ 未发现现有实验，无法继续")
        return False
    
    if existing['is_complete']:
        print("✅ 实验已完成，无需继续")
        return True
    
    # 分析当前进度
    try:
        current_df = pd.read_csv(existing['latest_intermediate'])
        completed_experiments = len(current_df)
        total_experiments = 25920
        progress = completed_experiments / total_experiments * 100
        
        print(f"📊 当前进度: {completed_experiments:,}/{total_experiments:,} ({progress:.1f}%)")
        print(f"⏱️ 剩余实验: {total_experiments - completed_experiments:,}")
        print(f"🕒 预计剩余时间: {(total_experiments - completed_experiments) * 0.5 / 3600:.1f} 小时")
        
        # 确认是否继续
        print("\\n❓ 是否继续完整实验？")
        print("   💡 可以在第10个cell中重新运行实验代码")
        print("   🔧 或者修改实验配置运行剩余部分")
        
        return False
        
    except Exception as e:
        print(f"❌ 读取进度文件失败: {e}")
        return False

def get_experiment_summary():
    """获取实验摘要信息"""
    existing = check_existing_results()
    if not existing['has_existing']:
        return None
    
    try:
        if existing['is_complete']:
            df = pd.read_csv(existing['complete_file'])
            status = "完成"
        else:
            df = pd.read_csv(existing['latest_intermediate'])
            status = "进行中"
        
        summary = {
            'status': status,
            'total_experiments': len(df),
            'success_rate': df['success'].mean() * 100,
            'avg_turns': df['turns_used'].mean(),
            'avg_api_calls': df['total_api_attempts'].mean(),
            'latest_time': df['timestamp'].iloc[-1],
            'file_path': existing['complete_file'] if existing['is_complete'] else existing['latest_intermediate']
        }
        
        return summary
    except Exception as e:
        print(f"❌ 获取摘要失败: {e}")
        return None

# 执行状态检查
print("🔄 检查实验继续状态...")
summary = get_experiment_summary()

if summary:
    print(f"\\n📋 实验摘要:")
    print(f"   状态: {summary['status']}")
    print(f"   已完成: {summary['total_experiments']:,} 个实验")
    print(f"   成功率: {summary['success_rate']:.1f}%")
    print(f"   平均轮数: {summary['avg_turns']:.1f}")
    print(f"   平均API调用: {summary['avg_api_calls']:.1f}")
    print(f"   最新时间: {summary['latest_time']}")
    print(f"   结果文件: {summary['file_path']}")
    
    if summary['status'] == "进行中":
        print(f"\\n🔄 实验可以继续")
        print(f"   📊 进度: {summary['total_experiments']/25920*100:.1f}%")
        print(f"   ⏰ 预计剩余: {(25920-summary['total_experiments'])*0.5/3600:.1f} 小时")
    else:
        print(f"\\n✅ 实验已完成！")
else:
    print("❌ 未发现有效的实验数据")

print("\\n💡 操作指南:")
print("   🔄 继续实验: 重新运行第10个cell")
print("   🧪 测试功能: 取消注释第12个cell的演示代码")
print("   📊 查看结果: 使用第11个cell的分析功能")
print("   🗑️ 重新开始: 删除results目录中的实验文件夹")
```
