```python
# OpenRouter API连通性测试
print("🌐 OpenRouter API连通性测试...")
print("=" * 60)

def test_openrouter_connectivity():
    """测试OpenRouter API的连通性"""
    
    # 获取API密钥
    openrouter_key = os.getenv('OPENROUTER_API_KEY')
    
    if not openrouter_key and os.path.exists('api_keys.json'):
        try:
            with open('api_keys.json', 'r') as f:
                keys = json.load(f)
                if 'openrouter_api_key' in keys and keys['openrouter_api_key'] != "your-openrouter-api-key-here":
                    openrouter_key = keys['openrouter_api_key']
        except Exception as e:
            print(f"⚠️ 读取api_keys.json失败: {e}")
    
    if not openrouter_key:
        print("❌ 未找到有效的OpenRouter API密钥")
        return {}
    
    print(f"🔍 开始测试OpenRouter API连通性...")
    
    # 选择代表性模型进行测试（普通模型和推理模型各选几个）
    test_models = [
        # 普通模型测试
        ('gpt-4o-mini', 'openai/gpt-4o-mini'),
        ('claude-3-7-sonnet', 'anthropic/claude-3.7-sonnet'),
        ('gemini-2-0-flash', 'google/gemini-2.0-flash'),
        # 推理模型测试  
        ('o1-mini', 'openai/o1-mini'),
        ('deepseek-r1', 'deepseek/deepseek-r1')
    ]
    
    connectivity_results = {}
    
    for short_name, model_name in test_models:
        print(f"\n🔄 测试 {short_name} ({model_name})...")
        try:
            # 准备请求
            headers = {
                'Authorization': f'Bearer {openrouter_key}',
                'Content-Type': 'application/json',
                'HTTP-Referer': 'https://github.com/taboo-benchmark',
                'X-Title': 'Taboo LLM Benchmark'
            }
            
            data = {
                'model': model_name,
                'messages': [{
                    'role': 'user',
                    'content': '测试连通性，请简短回复"连接成功"'
                }],
                'max_tokens': 50,
                'temperature': 0.1
            }
            
            # 发送请求
            response = requests.post(
                f"{OPENROUTER_CONFIG['base_url']}/chat/completions",
                headers=headers,
                json=data,
                timeout=30
            )
            
            if response.status_code == 200:
                result = response.json()
                if 'choices' in result and len(result['choices']) > 0:
                    content = result['choices'][0]['message']['content']
                    connectivity_results[short_name] = True
                    print(f"✅ {short_name} 连通性测试成功")
                    print(f"   响应: {content[:50]}...")
                else:
                    connectivity_results[short_name] = False
                    print(f"❌ {short_name} 连通性测试失败：响应格式异常")
            else:
                connectivity_results[short_name] = False
                print(f"❌ {short_name} 连通性测试失败：HTTP {response.status_code}")
                if response.status_code == 401:
                    print("   可能是API密钥无效")
                elif response.status_code == 429:
                    print("   请求频率过高，请稍后重试")
                
        except Exception as e:
            connectivity_results[short_name] = False
            print(f"❌ {short_name} 连通性测试失败: {e}")
    
    return connectivity_results

# 执行连通性测试
connectivity_results = test_openrouter_connectivity()

# 显示测试结果摘要
print(f"\n📊 OpenRouter连通性测试结果摘要:")
if connectivity_results:
    success_count = sum(connectivity_results.values())
    total_count = len(connectivity_results)
    
    for model, success in connectivity_results.items():
        status_icon = "✅" if success else "❌"
        print(f"   {status_icon} {model}: {'连通正常' if success else '连通失败'}")
    
    print(f"\n📈 总体连通率: {success_count}/{total_count} ({success_count/total_count:.1%})")
    
    if success_count > 0:
        print(f"✅ OpenRouter API工作正常，可以进行实验")
        print(f"🌟 通过OpenRouter可以访问 {len(OPENROUTER_CONFIG['models'])} 个不同的模型")
    else:
        print(f"❌ 所有模型连通性测试失败，请检查API密钥配置")
else:
    print("⚠️ 没有进行任何连通性测试（可能是因为没有配置API密钥）")

print(f"\n🔍 步骤2.2完成：OpenRouter API连通性测试完毕！")
```

```python
# 步骤3: OpenRouter实验配置
print("🧪 步骤3：OpenRouter模型实验配置...")
print("=" * 60)

# 检查前置条件
def check_openrouter_prerequisites():
    """检查OpenRouter实验前置条件"""
    print("🔍 检查实验前置条件...")
    
    # 检查数据集是否加载
    if 'experiment' not in globals():
        print("❌ 实验系统未初始化，请先运行步骤1")
        return False
    
    # 检查数据集
    if not experiment.datasets:
        print("❌ 数据集未加载，请先运行步骤1")
        return False
    
    # 检查OpenRouter连通性
    if 'connectivity_results' not in globals() or not connectivity_results:
        print("⚠️ 未进行OpenRouter连通性测试，请先运行步骤2")
        return False
    
    working_models = sum(connectivity_results.values())
    if working_models == 0:
        print("❌ 没有可用的OpenRouter模型连接，请检查步骤2的配置")
        return False
    
    print(f"✅ 前置条件检查通过")
    print(f"   - 数据集已加载: {len(experiment.datasets)} 个")
    print(f"   - 可用模型: {working_models} 个")
    
    return True

# OpenRouter实验配置
def configure_openrouter_experiment():
    """配置OpenRouter实验参数"""
    print("\\n⚙️ 配置OpenRouter实验参数...")
    
    # 获取可用的模型（基于连通性测试结果）
    available_models = []
    if connectivity_results:
        for model_name, is_available in connectivity_results.items():
            if is_available:
                available_models.append(model_name)
    
    # 如果连通性测试模型不够，添加更多OpenRouter模型
    if len(available_models) < 3:
        additional_models = ['gpt-4o', 'claude-3-5-sonnet', 'llama-3.1-70b']
        for model in additional_models:
            if model not in available_models:
                available_models.append(model)
    
    # 获取可用的数据集（优先WordNet）
    available_datasets = list(experiment.datasets.keys())
    priority_datasets = [d for d in available_datasets if d in ['wordnet', 'polysemous_wordnet']]
    if not priority_datasets:
        priority_datasets = available_datasets[:2]
    
    # OpenRouter实验配置
    openrouter_config = {
        'models': available_models[:4],  # 限制最多4个模型以控制成本
        'datasets': priority_datasets[:2],  # 最多2个数据集
        'games_per_config': 3,  # 每个配置的游戏数量
        'openrouter_models': OPENROUTER_CONFIG['models'],
        'base_url': OPENROUTER_CONFIG['base_url']
    }
    
    print(f"📊 OpenRouter实验配置:")
    print(f"   - 测试模型: {openrouter_config['models']}")
    print(f"   - 对应OpenRouter模型:")
    for model in openrouter_config['models']:
        if model in OPENROUTER_CONFIG['models']:
            print(f"     {model} -> {OPENROUTER_CONFIG['models'][model]}")
    print(f"   - 测试数据集: {openrouter_config['datasets']}")
    print(f"   - 每配置游戏数: {openrouter_config['games_per_config']}")
    
    total_configs = len(openrouter_config['models']) ** 2 * len(openrouter_config['datasets'])
    print(f"   - 总配置数: {total_configs}")
    print(f"   - 预计总游戏数: {total_configs * openrouter_config['games_per_config']}")
    
    return openrouter_config

# 检查前置条件
if check_openrouter_prerequisites():
    # 配置实验
    openrouter_exp_config = configure_openrouter_experiment()
    
    # 估算成本和时间
    total_games = len(openrouter_exp_config['models']) ** 2 * len(openrouter_exp_config['datasets']) * openrouter_exp_config['games_per_config']
    estimated_time = total_games * 30  # 假设每游戏30秒
    
    print(f"\\n💰 OpenRouter成本和时间估算:")
    print(f"   - 预计游戏数: {total_games}")
    print(f"   - 预计时间: {estimated_time//60:.0f} 分钟")
    print(f"   - 预计API调用数: {total_games * 6} 次")  # 每游戏约6次API调用
    print(f"   - OpenRouter优势: 统一计费，透明定价")
    
    print(f"\\n✅ OpenRouter实验配置完成，准备运行实验")
    
else:
    print("❌ 前置条件检查失败，请先完成前面的步骤")

print(f"\\n🔍 步骤3.1完成：OpenRouter实验配置就绪！")
```

```python
# 执行OpenRouter实验
print("🚀 执行OpenRouter实验...")
print("=" * 60)

# 实验执行控制
RUN_OPENROUTER_EXPERIMENT = False  # 设置为True以运行真实实验

def create_openrouter_llm_interface(model_name):
    """创建OpenRouter兼容的LLM接口"""
    
    # 获取API密钥
    openrouter_key = os.getenv('OPENROUTER_API_KEY')
    if not openrouter_key and os.path.exists('api_keys.json'):
        try:
            with open('api_keys.json', 'r') as f:
                keys = json.load(f)
                if 'openrouter_api_key' in keys:
                    openrouter_key = keys['openrouter_api_key']
        except:
            pass
    
    if not openrouter_key:
        raise ValueError("OpenRouter API密钥未配置")
    
    # 获取OpenRouter模型名称
    openrouter_model = OPENROUTER_CONFIG['models'].get(model_name, model_name)
    
    class OpenRouterLLM:
        def __init__(self, model_name, api_key):
            self.model_name = model_name
            self.api_key = api_key
            self.base_url = OPENROUTER_CONFIG['base_url']
        
        def generate_response(self, prompt, max_tokens=150, temperature=0.7):
            """生成响应"""
            headers = {
                'Authorization': f'Bearer {self.api_key}',
                'Content-Type': 'application/json',
                'HTTP-Referer': 'https://github.com/taboo-benchmark',
                'X-Title': 'Taboo LLM Benchmark'
            }
            
            data = {
                'model': self.model_name,
                'messages': [{'role': 'user', 'content': prompt}],
                'max_tokens': max_tokens,
                'temperature': temperature
            }
            
            try:
                response = requests.post(
                    f"{self.base_url}/chat/completions",
                    headers=headers,
                    json=data,
                    timeout=30
                )
                
                if response.status_code == 200:
                    result = response.json()
                    if 'choices' in result and len(result['choices']) > 0:
                        return result['choices'][0]['message']['content']
                    else:
                        return "响应格式错误"
                else:
                    return f"API错误: {response.status_code}"
                    
            except Exception as e:
                return f"请求失败: {e}"
    
    return OpenRouterLLM(openrouter_model, openrouter_key)

def run_openrouter_experiment():
    """运行OpenRouter实验"""
    if 'openrouter_exp_config' not in globals():
        print("❌ OpenRouter实验配置未完成，请先运行上一个单元格")
        return None
    
    print("🎯 开始运行OpenRouter实验...")
    print(f"📊 实验配置: {len(openrouter_exp_config['models'])} 个模型 × {len(openrouter_exp_config['datasets'])} 个数据集")
    
    results_summary = []
    total_configs = len(openrouter_exp_config['models']) ** 2 * len(openrouter_exp_config['datasets'])
    current_config = 0
    
    start_time = time.time()
    
    # 运行所有配置组合
    for hinter_model in openrouter_exp_config['models']:
        for guesser_model in openrouter_exp_config['models']:
            for dataset in openrouter_exp_config['datasets']:
                current_config += 1
                print(f"\\n🔄 进度: {current_config}/{total_configs} - {hinter_model} → {guesser_model} on {dataset}")
                
                try:
                    config_start_time = time.time()
                    
                    # 创建OpenRouter LLM接口
                    hinter_llm = create_openrouter_llm_interface(hinter_model)
                    guesser_llm = create_openrouter_llm_interface(guesser_model)
                    
                    # 运行少量游戏进行测试
                    games_played = 0
                    successful_games = 0
                    total_turns = 0
                    
                    # 从数据集中选择几个词进行测试
                    dataset_items = experiment.datasets[dataset]
                    test_items = dataset_items.sample(min(openrouter_exp_config['games_per_config'], len(dataset_items)))
                    
                    for item in test_items:
                        games_played += 1
                        
                        # 模拟游戏结果（简化版）
                        target_word = item.get('target', 'test')
                        taboo_words = item.get('taboo', [])
                        
                        # 简单的测试：让hinter描述词汇
                        hinter_prompt = f"请用简短的句子描述'{target_word}'，但不能使用这些词: {', '.join(taboo_words[:3])}"
                        hinter_response = hinter_llm.generate_response(hinter_prompt, max_tokens=50)
                        
                        # 让guesser猜测
                        guesser_prompt = f"根据这个描述，猜测目标词汇是什么：{hinter_response}"
                        guesser_response = guesser_llm.generate_response(guesser_prompt, max_tokens=20)
                        
                        # 简单判断是否成功（包含目标词）
                        if target_word.lower() in guesser_response.lower():
                            successful_games += 1
                            total_turns += 1
                        else:
                            total_turns += 2
                    
                    config_end_time = time.time()
                    config_duration = config_end_time - config_start_time
                    
                    result_summary = {
                        'hinter_model': hinter_model,
                        'guesser_model': guesser_model,
                        'dataset': dataset,
                        'success_rate': successful_games / games_played if games_played > 0 else 0,
                        'total_games': games_played,
                        'avg_turns': total_turns / games_played if games_played > 0 else 0,
                        'duration_seconds': config_duration
                    }
                    results_summary.append(result_summary)
                    
                    print(f"   ✅ 完成 - 成功率: {result_summary['success_rate']:.1%}, 用时: {config_duration:.1f}秒")
                    
                except Exception as e:
                    print(f"   ❌ 失败: {e}")
                    continue
    
    end_time = time.time()
    total_duration = end_time - start_time
    
    if results_summary:
        # 创建结果数据框
        results_df = pd.DataFrame(results_summary)
        
        # 保存结果
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        results_file = f"openrouter_experiment_results_{timestamp}.csv"
        results_df.to_csv(results_file, index=False)
        
        print(f"\\n📊 OpenRouter实验完成摘要:")
        print(f"   - 成功配置数: {len(results_df)}")
        print(f"   - 总用时: {total_duration:.1f} 秒 ({total_duration/60:.1f} 分钟)")
        print(f"   - 平均每配置用时: {total_duration/len(results_df):.1f} 秒")
        print(f"   - 结果文件: {results_file}")
        
        return results_df
    else:
        print("\\n⚠️ 没有成功完成任何实验配置")
        return None

# 实验执行
if RUN_OPENROUTER_EXPERIMENT:
    if 'openrouter_exp_config' in globals():
        print("⚠️ 即将开始OpenRouter真实实验，这将产生API调用费用")
        print("💰 预计成本:")
        
        total_games = len(openrouter_exp_config['models']) ** 2 * len(openrouter_exp_config['datasets']) * openrouter_exp_config['games_per_config']
        api_calls = total_games * 4  # 每游戏约4次API调用（简化版）
        
        print(f"   - 预计API调用数: {api_calls}")
        print(f"   - OpenRouter统一计费，费用透明")
        
        print("\\n🚀 开始OpenRouter实验...")
        openrouter_experiment_results = run_openrouter_experiment()
        
        if openrouter_experiment_results is not None:
            print("\\n✅ OpenRouter实验执行完成！")
        else:
            print("\\n❌ OpenRouter实验执行失败")
    else:
        print("❌ 请先运行OpenRouter实验配置")
        
else:
    print("💡 要运行OpenRouter真实实验，请执行以下步骤:")
    print("1. 确保已完成步骤1和步骤2")
    print("2. 设置 RUN_OPENROUTER_EXPERIMENT = True")
    print("3. 重新运行此单元格")
    print("\\n⚠️ 注意：运行真实实验将产生OpenRouter API调用费用")
    print("💡 建议先进行小规模测试（减少games_per_config的值）")

print(f"\\n🔍 步骤3.2完成：OpenRouter实验执行控制就绪！")
```

```python
# OpenRouter模型性能分析
print("📈 OpenRouter模型性能分析...")
print("=" * 60)

# 按提供商分析模型性能
def analyze_openrouter_performance(df):
    """分析OpenRouter模型性能，按提供商分组"""
    
    # 添加提供商信息
    df_analysis = df.copy()
    
    def get_provider(model_name):
        openrouter_model = OPENROUTER_CONFIG['models'].get(model_name, model_name)
        return openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    
    df_analysis['hinter_provider'] = df_analysis['hinter_model'].apply(get_provider)
    df_analysis['guesser_provider'] = df_analysis['guesser_model'].apply(get_provider)
    
    # 根据数据结构调整分析逻辑
    if 'success' in df_analysis.columns:
        # 个别游戏数据
        hinter_performance = df_analysis.groupby('hinter_model')['success'].agg(['mean', 'count']).round(3)
        hinter_performance.columns = ['成功率', '游戏数']
        
        guesser_performance = df_analysis.groupby('guesser_model')['success'].agg(['mean', 'count']).round(3)
        guesser_performance.columns = ['成功率', '游戏数']
        
        # 按提供商分析
        provider_hinter_perf = df_analysis.groupby('hinter_provider')['success'].agg(['mean', 'count']).round(3)
        provider_hinter_perf.columns = ['成功率', '游戏数']
        
        provider_guesser_perf = df_analysis.groupby('guesser_provider')['success'].agg(['mean', 'count']).round(3)
        provider_guesser_perf.columns = ['成功率', '游戏数']
        
    else:
        # 聚合实验结果数据
        if 'success_rate' in df_analysis.columns:
            hinter_performance = df_analysis.groupby('hinter_model')['success_rate'].agg(['mean', 'count']).round(3)
            hinter_performance.columns = ['平均成功率', '配置数']
            
            guesser_performance = df_analysis.groupby('guesser_model')['success_rate'].agg(['mean', 'count']).round(3)
            guesser_performance.columns = ['平均成功率', '配置数']
            
            provider_hinter_perf = df_analysis.groupby('hinter_provider')['success_rate'].agg(['mean', 'count']).round(3)
            provider_hinter_perf.columns = ['平均成功率', '配置数']
            
            provider_guesser_perf = df_analysis.groupby('guesser_provider')['success_rate'].agg(['mean', 'count']).round(3)
            provider_guesser_perf.columns = ['平均成功率', '配置数']
    
    return hinter_performance, guesser_performance, provider_hinter_perf, provider_guesser_perf

# 执行性能分析
hinter_perf, guesser_perf, provider_hinter_perf, provider_guesser_perf = analyze_openrouter_performance(df)

print("🎯 提示者(Hinter)模型性能:")
for model, row in hinter_perf.sort_values(hinter_perf.columns[0], ascending=False).iterrows():
    openrouter_model = OPENROUTER_CONFIG['models'].get(model, model)
    provider = openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    print(f"   {model} ({provider}): {row.iloc[0]:.1%} ({row.iloc[1]} 次)")

print("\\n🔍 猜测者(Guesser)模型性能:")
for model, row in guesser_perf.sort_values(guesser_perf.columns[0], ascending=False).iterrows():
    openrouter_model = OPENROUTER_CONFIG['models'].get(model, model)
    provider = openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    print(f"   {model} ({provider}): {row.iloc[0]:.1%} ({row.iloc[1]} 次)")

print("\\n🏢 按提供商分析 - 提示者性能:")
for provider, row in provider_hinter_perf.sort_values(provider_hinter_perf.columns[0], ascending=False).iterrows():
    print(f"   {provider}: {row.iloc[0]:.1%} ({row.iloc[1]} 次)")

print("\\n🏢 按提供商分析 - 猜测者性能:")
for provider, row in provider_guesser_perf.sort_values(provider_guesser_perf.columns[0], ascending=False).iterrows():
    print(f"   {provider}: {row.iloc[0]:.1%} ({row.iloc[1]} 次)")

# 数据集难度分析
if 'success' in df.columns:
    dataset_performance = df.groupby('dataset')['success'].agg(['mean', 'count']).round(3)
    dataset_performance.columns = ['成功率', '游戏数']
elif 'success_rate' in df.columns:
    dataset_performance = df.groupby('dataset')['success_rate'].agg(['mean', 'count']).round(3)
    dataset_performance.columns = ['平均成功率', '配置数']

print("\\n📊 数据集难度分析:")
for dataset, row in dataset_performance.sort_values(dataset_performance.columns[0], ascending=False).iterrows():
    print(f"   {dataset}: {row.iloc[0]:.1%} ({row.iloc[1]} 次)")

# 模型组合分析
if 'success' in df.columns:
    print("\\n🤝 最佳OpenRouter模型组合 (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['成功率', '游戏数']
    top_combinations = combination_perf.sort_values('成功率', ascending=False).head()
    for (hinter, guesser), row in top_combinations.iterrows():
        hinter_provider = OPENROUTER_CONFIG['models'].get(hinter, hinter).split('/')[0]
        guesser_provider = OPENROUTER_CONFIG['models'].get(guesser, guesser).split('/')[0]
        print(f"   {hinter} ({hinter_provider}) → {guesser} ({guesser_provider}): {row['成功率']:.1%}")
elif 'success_rate' in df.columns:
    print("\\n🤝 最佳OpenRouter模型组合 (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success_rate'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['平均成功率', '配置数']
    top_combinations = combination_perf.sort_values('平均成功率', ascending=False).head()
    for (hinter, guesser), row in top_combinations.iterrows():
        hinter_provider = OPENROUTER_CONFIG['models'].get(hinter, hinter).split('/')[0]
        guesser_provider = OPENROUTER_CONFIG['models'].get(guesser, guesser).split('/')[0]
        print(f"   {hinter} ({hinter_provider}) → {guesser} ({guesser_provider}): {row['平均成功率']:.1%}")

print(f"\\n🔍 步骤4.2完成：OpenRouter性能分析完毕！")
```

```python
# 步骤2: API密钥配置
print("🔐 步骤2：API密钥配置和连通性测试...")
print("=" * 60)

# API密钥配置功能
def configure_api_keys():
    """配置API密钥"""
    print("📝 API密钥配置选项：")
    print("1. 通过环境变量配置（推荐）")
    print("2. 通过api_keys.json文件配置")
    print("3. 在notebook中直接设置（不推荐，安全性低）")
    
    # 检查当前API密钥状态
    api_status = {}
    api_status['openai'] = bool(os.getenv('OPENAI_API_KEY'))
    api_status['anthropic'] = bool(os.getenv('ANTHROPIC_API_KEY')) 
    api_status['google'] = bool(os.getenv('GOOGLE_API_KEY'))
    
    # 检查api_keys.json文件
    if os.path.exists('api_keys.json'):
        try:
            with open('api_keys.json', 'r') as f:
                keys = json.load(f)
                if not api_status['openai'] and 'openai_api_key' in keys:
                    api_status['openai'] = bool(keys['openai_api_key'])
                if not api_status['anthropic'] and 'anthropic_api_key' in keys:
                    api_status['anthropic'] = bool(keys['anthropic_api_key'])
                if not api_status['google'] and 'google_api_key' in keys:
                    api_status['google'] = bool(keys['google_api_key'])
        except:
            pass
    
    print(f"\n📊 当前API密钥状态:")
    for provider, status in api_status.items():
        status_icon = "✅" if status else "❌"
        print(f"   {status_icon} {provider.upper()}: {'已配置' if status else '未配置'}")
    
    return api_status

def create_api_keys_template():
    """创建API密钥配置模板"""
    template = {
        "openai_api_key": "your-openai-api-key-here",
        "anthropic_api_key": "your-anthropic-api-key-here", 
        "google_api_key": "your-google-api-key-here"
    }
    
    if not os.path.exists('api_keys.json'):
        with open('api_keys.json', 'w') as f:
            json.dump(template, f, indent=2)
        print("📄 已创建api_keys.json模板文件")
        print("💡 请编辑此文件并填入您的API密钥")
    else:
        print("📄 api_keys.json文件已存在")

# 配置API密钥
api_status = configure_api_keys()

# 创建API密钥模板文件
create_api_keys_template()

print(f"\n📖 配置说明:")
print(f"方法1 - 环境变量（推荐）:")
print(f"   export OPENAI_API_KEY='your-key'")
print(f"   export ANTHROPIC_API_KEY='your-key'")
print(f"   export GOOGLE_API_KEY='your-key'")
print(f"\n方法2 - 编辑api_keys.json文件:")
print(f"   在项目根目录的api_keys.json中填入您的密钥")

# 检查是否至少有一个API密钥配置
has_any_key = any(api_status.values())
if has_any_key:
    print(f"\n✅ 至少配置了一个API密钥，可以进行连通性测试")
else:
    print(f"\n⚠️ 未检测到任何API密钥配置")
    print(f"💡 请先配置API密钥后重新运行此单元格")

print(f"\n🔍 步骤2.1完成：API密钥配置检查完毕！")
```

```python
# API连通性测试
print("🌐 API连通性测试...")
print("=" * 60)

def test_api_connectivity():
    """测试各个API的连通性"""
    
    # 重新读取API密钥状态
    api_keys = {}
    
    # 从环境变量读取
    if os.getenv('OPENAI_API_KEY'):
        api_keys['openai'] = os.getenv('OPENAI_API_KEY')
    if os.getenv('ANTHROPIC_API_KEY'):
        api_keys['anthropic'] = os.getenv('ANTHROPIC_API_KEY')
    if os.getenv('GOOGLE_API_KEY'):
        api_keys['google'] = os.getenv('GOOGLE_API_KEY')
    
    # 从api_keys.json读取
    if os.path.exists('api_keys.json'):
        try:
            with open('api_keys.json', 'r') as f:
                file_keys = json.load(f)
                if 'openai_api_key' in file_keys and file_keys['openai_api_key'] != "your-openai-api-key-here":
                    api_keys['openai'] = file_keys['openai_api_key']
                if 'anthropic_api_key' in file_keys and file_keys['anthropic_api_key'] != "your-anthropic-api-key-here":
                    api_keys['anthropic'] = file_keys['anthropic_api_key']
                if 'google_api_key' in file_keys and file_keys['google_api_key'] != "your-google-api-key-here":
                    api_keys['google'] = file_keys['google_api_key']
        except Exception as e:
            print(f"⚠️ 读取api_keys.json失败: {e}")
    
    if not api_keys:
        print("❌ 未找到有效的API密钥")
        return {}
    
    print(f"🔍 开始测试 {len(api_keys)} 个API的连通性...")
    
    connectivity_results = {}
    
    # 测试OpenAI API
    if 'openai' in api_keys:
        print("\n🔄 测试OpenAI API...")
        try:
            # 临时设置环境变量用于测试
            original_key = os.getenv('OPENAI_API_KEY')
            os.environ['OPENAI_API_KEY'] = api_keys['openai']
            
            # 创建LLM接口进行测试
            from llm_interface import create_llm_interface
            llm = create_llm_interface("gpt-4o-mini", mock=False)
            
            # 进行简单的API调用测试
            test_response = llm.generate_response("测试连通性，请回复'连接成功'")
            
            if test_response and len(test_response) > 0:
                connectivity_results['openai'] = True
                print("✅ OpenAI API连通性测试成功")
            else:
                connectivity_results['openai'] = False
                print("❌ OpenAI API连通性测试失败：响应为空")
                
            # 恢复原始环境变量
            if original_key:
                os.environ['OPENAI_API_KEY'] = original_key
            elif 'OPENAI_API_KEY' in os.environ:
                del os.environ['OPENAI_API_KEY']
                
        except Exception as e:
            connectivity_results['openai'] = False
            print(f"❌ OpenAI API连通性测试失败: {e}")
    
    # 测试Anthropic API
    if 'anthropic' in api_keys:
        print("\n🔄 测试Anthropic API...")
        try:
            original_key = os.getenv('ANTHROPIC_API_KEY')
            os.environ['ANTHROPIC_API_KEY'] = api_keys['anthropic']
            
            llm = create_llm_interface("claude-3-5-sonnet-20241022", mock=False)
            test_response = llm.generate_response("测试连通性，请回复'连接成功'")
            
            if test_response and len(test_response) > 0:
                connectivity_results['anthropic'] = True
                print("✅ Anthropic API连通性测试成功")
            else:
                connectivity_results['anthropic'] = False
                print("❌ Anthropic API连通性测试失败：响应为空")
                
            if original_key:
                os.environ['ANTHROPIC_API_KEY'] = original_key
            elif 'ANTHROPIC_API_KEY' in os.environ:
                del os.environ['ANTHROPIC_API_KEY']
                
        except Exception as e:
            connectivity_results['anthropic'] = False
            print(f"❌ Anthropic API连通性测试失败: {e}")
    
    # 测试Google API
    if 'google' in api_keys:
        print("\n🔄 测试Google API...")
        try:
            original_key = os.getenv('GOOGLE_API_KEY')
            os.environ['GOOGLE_API_KEY'] = api_keys['google']
            
            llm = create_llm_interface("gemini-1.5-flash", mock=False)
            test_response = llm.generate_response("测试连通性，请回复'连接成功'")
            
            if test_response and len(test_response) > 0:
                connectivity_results['google'] = True
                print("✅ Google API连通性测试成功")
            else:
                connectivity_results['google'] = False
                print("❌ Google API连通性测试失败：响应为空")
                
            if original_key:
                os.environ['GOOGLE_API_KEY'] = original_key
            elif 'GOOGLE_API_KEY' in os.environ:
                del os.environ['GOOGLE_API_KEY']
                
        except Exception as e:
            connectivity_results['google'] = False
            print(f"❌ Google API连通性测试失败: {e}")
    
    return connectivity_results

# 执行连通性测试
connectivity_results = test_api_connectivity()

# 显示测试结果摘要
print(f"\n📊 连通性测试结果摘要:")
if connectivity_results:
    success_count = sum(connectivity_results.values())
    total_count = len(connectivity_results)
    
    for provider, success in connectivity_results.items():
        status_icon = "✅" if success else "❌"
        print(f"   {status_icon} {provider.upper()}: {'连通正常' if success else '连通失败'}")
    
    print(f"\n📈 总体连通率: {success_count}/{total_count} ({success_count/total_count:.1%})")
    
    if success_count > 0:
        print(f"✅ 有 {success_count} 个API可用，可以进行实验")
    else:
        print(f"❌ 所有API连通性测试失败，请检查密钥配置")
else:
    print("⚠️ 没有进行任何连通性测试（可能是因为没有配置API密钥）")

print(f"\n🔍 步骤2.2完成：API连通性测试完毕！")
```

```python
# 步骤3: 实验配置和执行
print("🧪 步骤3：实验配置和执行...")
print("=" * 60)

# 检查前置条件
def check_prerequisites():
    """检查实验前置条件"""
    print("🔍 检查实验前置条件...")
    
    # 检查数据集是否加载
    if 'experiment' not in globals():
        print("❌ 实验系统未初始化，请先运行步骤1")
        return False
    
    # 检查数据集
    if not experiment.datasets:
        print("❌ 数据集未加载，请先运行步骤1")
        return False
    
    # 检查API连通性
    if 'connectivity_results' not in globals() or not connectivity_results:
        print("⚠️ 未进行API连通性测试，请先运行步骤2")
        return False
    
    working_apis = sum(connectivity_results.values())
    if working_apis == 0:
        print("❌ 没有可用的API连接，请检查步骤2的配置")
        return False
    
    print(f"✅ 前置条件检查通过")
    print(f"   - 数据集已加载: {len(experiment.datasets)} 个")
    print(f"   - 可用API: {working_apis} 个")
    
    return True

# 实验配置
def configure_experiment():
    """配置实验参数"""
    print("\n⚙️ 配置实验参数...")
    
    # 获取可用的模型（基于连通性测试结果）
    available_models = []
    if connectivity_results.get('openai', False):
        available_models.extend(["gpt-4o-mini", "gpt-4o"])
    if connectivity_results.get('anthropic', False):
        available_models.extend(["claude-3-5-sonnet-20241022"])
    if connectivity_results.get('google', False):
        available_models.extend(["gemini-1.5-flash", "gemini-1.5-pro"])
    
    # 获取可用的数据集
    available_datasets = list(experiment.datasets.keys())
    
    # 实验配置
    experiment_config = {
        'models': available_models[:3],  # 限制最多3个模型以控制成本
        'datasets': [d for d in available_datasets if d in ['wordnet', 'polysemous_wordnet']][:2],  # 优先WordNet数据集
        'games_per_config': 3,  # 每个配置的游戏数量
        'max_experiments': 18   # 最大实验数量限制
    }
    
    # 如果没有WordNet数据集，使用其他可用数据集
    if not experiment_config['datasets']:
        experiment_config['datasets'] = available_datasets[:2]
    
    print(f"📊 实验配置:")
    print(f"   - 测试模型: {experiment_config['models']}")
    print(f"   - 测试数据集: {experiment_config['datasets']}")
    print(f"   - 每配置游戏数: {experiment_config['games_per_config']}")
    
    total_configs = len(experiment_config['models']) * len(experiment_config['models']) * len(experiment_config['datasets'])
    print(f"   - 总配置数: {total_configs}")
    print(f"   - 预计总游戏数: {total_configs * experiment_config['games_per_config']}")
    
    return experiment_config

# 检查前置条件
if check_prerequisites():
    # 配置实验
    exp_config = configure_experiment()
    
    # 估算成本和时间
    total_games = len(exp_config['models']) ** 2 * len(exp_config['datasets']) * exp_config['games_per_config']
    estimated_time = total_games * 30  # 假设每游戏30秒
    
    print(f"\n💰 成本和时间估算:")
    print(f"   - 预计游戏数: {total_games}")
    print(f"   - 预计时间: {estimated_time//60:.0f} 分钟")
    print(f"   - 预计API调用数: {total_games * 6} 次")  # 每游戏约6次API调用
    
    print(f"\n✅ 实验配置完成，准备运行实验")
    
else:
    print("❌ 前置条件检查失败，请先完成前面的步骤")

print(f"\n🔍 步骤3.1完成：实验配置就绪！")
```

```python
# 执行实验
print("🚀 执行实验...")
print("=" * 60)

# 实验执行控制
RUN_EXPERIMENT = False  # 设置为True以运行真实实验

def run_wordnet_experiment():
    """运行WordNet实验"""
    if 'exp_config' not in globals():
        print("❌ 实验配置未完成，请先运行上一个单元格")
        return None
    
    print("🎯 开始运行WordNet实验...")
    print(f"📊 实验配置: {len(exp_config['models'])} 个模型 × {len(exp_config['datasets'])} 个数据集")
    
    results_summary = []
    total_configs = len(exp_config['models']) ** 2 * len(exp_config['datasets'])
    current_config = 0
    
    start_time = time.time()
    
    # 运行所有配置组合
    for hinter in exp_config['models']:
        for guesser in exp_config['models']:
            for dataset in exp_config['datasets']:
                current_config += 1
                print(f"\n🔄 进度: {current_config}/{total_configs} - {hinter} → {guesser} on {dataset}")
                
                try:
                    config_start_time = time.time()
                    
                    # 运行单个配置
                    result = experiment.run_single_configuration(
                        hinter_model=hinter,
                        guesser_model=guesser,
                        dataset_name=dataset
                    )
                    
                    config_end_time = time.time()
                    config_duration = config_end_time - config_start_time
                    
                    result_summary = {
                        'hinter_model': hinter,
                        'guesser_model': guesser,
                        'dataset': dataset,
                        'success_rate': result['success_rate'],
                        'total_games': result['total_games'],
                        'avg_turns': result['avg_turns'],
                        'duration_seconds': config_duration
                    }
                    results_summary.append(result_summary)
                    
                    print(f"   ✅ 完成 - 成功率: {result['success_rate']:.1%}, 用时: {config_duration:.1f}秒")
                    
                except Exception as e:
                    print(f"   ❌ 失败: {e}")
                    continue
    
    end_time = time.time()
    total_duration = end_time - start_time
    
    if results_summary:
        # 创建结果数据框
        results_df = pd.DataFrame(results_summary)
        
        # 保存结果
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        results_file = f"wordnet_experiment_results_{timestamp}.csv"
        results_df.to_csv(results_file, index=False)
        
        print(f"\n📊 实验完成摘要:")
        print(f"   - 成功配置数: {len(results_df)}")
        print(f"   - 总用时: {total_duration:.1f} 秒 ({total_duration/60:.1f} 分钟)")
        print(f"   - 平均每配置用时: {total_duration/len(results_df):.1f} 秒")
        print(f"   - 结果文件: {results_file}")
        
        return results_df
    else:
        print("\n⚠️ 没有成功完成任何实验配置")
        return None

# 实验执行
if RUN_EXPERIMENT:
    if 'exp_config' in globals():
        print("⚠️ 即将开始真实实验，这将产生API调用费用")
        print("💰 预计成本:")
        
        total_games = len(exp_config['models']) ** 2 * len(exp_config['datasets']) * exp_config['games_per_config']
        api_calls = total_games * 6  # 每游戏约6次API调用
        
        # 粗略成本估算（基于OpenAI定价）
        estimated_cost = api_calls * 0.0001  # 假设每次调用0.0001美元
        print(f"   - 预计API调用数: {api_calls}")
        print(f"   - 预计成本: ${estimated_cost:.2f} USD")
        
        print("\n🚀 开始实验...")
        experiment_results = run_wordnet_experiment()
        
        if experiment_results is not None:
            print("\n✅ 实验执行完成！")
        else:
            print("\n❌ 实验执行失败")
    else:
        print("❌ 请先运行实验配置")
        
else:
    print("💡 要运行真实实验，请执行以下步骤:")
    print("1. 确保已完成步骤1和步骤2")
    print("2. 设置 RUN_EXPERIMENT = True")
    print("3. 重新运行此单元格")
    print("\n⚠️ 注意：运行真实实验将产生API调用费用")
    print("💡 建议先进行小规模测试（减少games_per_config的值）")

print(f"\n🔍 步骤3.2完成：实验执行控制就绪！")
```

```python
# 模型性能比较分析
print("📈 模型性能比较分析...")
print("=" * 60)

# 计算性能指标
def analyze_model_performance(df):
    """分析模型性能"""
    
    # 根据数据结构调整分析逻辑
    if 'success' in df.columns:
        # 个别游戏数据
        hinter_performance = df.groupby('hinter_model')['success'].agg(['mean', 'count']).round(3)
        hinter_performance.columns = ['成功率', '游戏数']
        
        guesser_performance = df.groupby('guesser_model')['success'].agg(['mean', 'count']).round(3)
        guesser_performance.columns = ['成功率', '游戏数']
        
        dataset_performance = df.groupby('dataset')['success'].agg(['mean', 'count']).round(3)
        dataset_performance.columns = ['成功率', '游戏数']
        
    else:
        # 聚合实验结果数据
        if 'success_rate' in df.columns:
            hinter_performance = df.groupby('hinter_model')['success_rate'].agg(['mean', 'count']).round(3)
            hinter_performance.columns = ['平均成功率', '配置数']
            
            guesser_performance = df.groupby('guesser_model')['success_rate'].agg(['mean', 'count']).round(3)
            guesser_performance.columns = ['平均成功率', '配置数']
            
            dataset_performance = df.groupby('dataset')['success_rate'].agg(['mean', 'count']).round(3)
            dataset_performance.columns = ['平均成功率', '配置数']
    
    return hinter_performance, guesser_performance, dataset_performance

# 执行性能分析
hinter_perf, guesser_perf, dataset_perf = analyze_model_performance(df)

print("🎯 提示者(Hinter)模型性能:")
print(hinter_perf.sort_values(hinter_perf.columns[0], ascending=False))

print("\n🔍 猜测者(Guesser)模型性能:")
print(guesser_perf.sort_values(guesser_perf.columns[0], ascending=False))

print("\n📊 数据集难度分析:")
print(dataset_perf.sort_values(dataset_perf.columns[0], ascending=False))

# 模型组合分析
if 'success' in df.columns:
    print("\n🤝 最佳模型组合 (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['成功率', '游戏数']
    top_combinations = combination_perf.sort_values('成功率', ascending=False).head()
    print(top_combinations)
elif 'success_rate' in df.columns:
    print("\n🤝 最佳模型组合 (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success_rate'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['平均成功率', '配置数']
    top_combinations = combination_perf.sort_values('平均成功率', ascending=False).head()
    print(top_combinations)

print(f"\n🔍 步骤4.2完成：性能分析完毕！")
```

```python
# 详细结果报告和导出
print("📄 详细结果报告和导出...")
print("=" * 60)

# 生成详细报告
def generate_detailed_report(df):
    """生成详细的实验报告"""
    
    report = {
        'experiment_summary': {},
        'model_analysis': {},
        'dataset_analysis': {},
        'recommendations': []
    }
    
    # 实验摘要
    report['experiment_summary'] = {
        'total_records': len(df),
        'unique_models': len(df['hinter_model'].unique()),
        'unique_datasets': len(df['dataset'].unique()),
        'data_source': data_source
    }
    
    if 'success' in df.columns:
        report['experiment_summary']['overall_success_rate'] = f"{df['success'].mean():.1%}"
        report['experiment_summary']['total_games'] = len(df)
    elif 'success_rate' in df.columns:
        report['experiment_summary']['avg_success_rate'] = f"{df['success_rate'].mean():.1%}"
        if 'total_games' in df.columns:
            report['experiment_summary']['total_games'] = df['total_games'].sum()
    
    # 模型分析
    if 'success' in df.columns:
        hinter_perf = df.groupby('hinter_model')['success'].mean().sort_values(ascending=False)
        guesser_perf = df.groupby('guesser_model')['success'].mean().sort_values(ascending=False)
    else:
        hinter_perf = df.groupby('hinter_model')['success_rate'].mean().sort_values(ascending=False)
        guesser_perf = df.groupby('guesser_model')['success_rate'].mean().sort_values(ascending=False)
    
    report['model_analysis'] = {
        'best_hinter': hinter_perf.index[0],
        'best_hinter_score': f"{hinter_perf.iloc[0]:.1%}",
        'best_guesser': guesser_perf.index[0],
        'best_guesser_score': f"{guesser_perf.iloc[0]:.1%}",
        'hinter_ranking': hinter_perf.to_dict(),
        'guesser_ranking': guesser_perf.to_dict()
    }
    
    # 数据集分析
    if 'success' in df.columns:
        dataset_perf = df.groupby('dataset')['success'].mean().sort_values(ascending=False)
    else:
        dataset_perf = df.groupby('dataset')['success_rate'].mean().sort_values(ascending=False)
    
    report['dataset_analysis'] = {
        'easiest_dataset': dataset_perf.index[0],
        'easiest_score': f"{dataset_perf.iloc[0]:.1%}",
        'hardest_dataset': dataset_perf.index[-1],
        'hardest_score': f"{dataset_perf.iloc[-1]:.1%}",
        'dataset_ranking': dataset_perf.to_dict()
    }
    
    # 生成建议
    recommendations = []
    
    if report['model_analysis']['best_hinter_score'] > "75%":
        recommendations.append(f"推荐使用 {report['model_analysis']['best_hinter']} 作为提示者模型")
    
    if report['model_analysis']['best_guesser_score'] > "75%":
        recommendations.append(f"推荐使用 {report['model_analysis']['best_guesser']} 作为猜测者模型")
    
    if report['dataset_analysis']['easiest_score'] > "80%":
        recommendations.append(f"{report['dataset_analysis']['easiest_dataset']} 数据集适合模型训练和评估")
    
    if report['dataset_analysis']['hardest_score'] < "60%":
        recommendations.append(f"{report['dataset_analysis']['hardest_dataset']} 数据集具有挑战性，适合高级模型测试")
    
    report['recommendations'] = recommendations
    
    return report

# 生成报告
detailed_report = generate_detailed_report(df)

print("📊 实验总结报告")
print(f"   - 数据来源: {detailed_report['experiment_summary']['data_source']}")
print(f"   - 总记录数: {detailed_report['experiment_summary']['total_records']}")
print(f"   - 测试模型数: {detailed_report['experiment_summary']['unique_models']}")
print(f"   - 测试数据集数: {detailed_report['experiment_summary']['unique_datasets']}")

if 'overall_success_rate' in detailed_report['experiment_summary']:
    print(f"   - 总体成功率: {detailed_report['experiment_summary']['overall_success_rate']}")
elif 'avg_success_rate' in detailed_report['experiment_summary']:
    print(f"   - 平均成功率: {detailed_report['experiment_summary']['avg_success_rate']}")

print(f"\n🏆 最佳模型表现:")
print(f"   - 最佳提示者: {detailed_report['model_analysis']['best_hinter']} ({detailed_report['model_analysis']['best_hinter_score']})")
print(f"   - 最佳猜测者: {detailed_report['model_analysis']['best_guesser']} ({detailed_report['model_analysis']['best_guesser_score']})")

print(f"\n📊 数据集难度:")
print(f"   - 最容易: {detailed_report['dataset_analysis']['easiest_dataset']} ({detailed_report['dataset_analysis']['easiest_score']})")
print(f"   - 最困难: {detailed_report['dataset_analysis']['hardest_dataset']} ({detailed_report['dataset_analysis']['hardest_score']})")

print(f"\n💡 建议:")
for i, recommendation in enumerate(detailed_report['recommendations'], 1):
    print(f"   {i}. {recommendation}")

# 导出结果
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

# 导出CSV
csv_filename = f"taboo_benchmark_results_{timestamp}.csv"
df.to_csv(csv_filename, index=False, encoding='utf-8')
print(f"\n📁 结果已导出:")
print(f"   - CSV文件: {csv_filename}")

# 导出Excel（如果数据不太大）
if len(df) < 10000:
    try:
        excel_filename = f"taboo_benchmark_results_{timestamp}.xlsx"
        with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name='原始数据', index=False)
            
            # 添加汇总工作表
            if 'success' in df.columns:
                summary_data = {
                    '模型': list(hinter_perf.index),
                    '提示者成功率': [f"{x:.1%}" for x in hinter_perf.values],
                    '猜测者成功率': [f"{guesser_perf.loc[model]:.1%}" for model in hinter_perf.index]
                }
            else:
                summary_data = {
                    '模型': list(hinter_perf.index),
                    '提示者平均成功率': [f"{x:.1%}" for x in hinter_perf.values],
                    '猜测者平均成功率': [f"{guesser_perf.loc[model]:.1%}" for model in hinter_perf.index]
                }
            
            summary_df = pd.DataFrame(summary_data)
            summary_df.to_excel(writer, sheet_name='模型性能汇总', index=False)
        
        print(f"   - Excel文件: {excel_filename}")
    except Exception as e:
        print(f"   - Excel导出失败: {e}")

# 保存详细报告
report_filename = f"taboo_benchmark_report_{timestamp}.json"
with open(report_filename, 'w', encoding='utf-8') as f:
    import json
    json.dump(detailed_report, f, ensure_ascii=False, indent=2)

print(f"   - 详细报告: {report_filename}")

print(f"\n🔍 步骤4.4完成：报告生成和导出完毕！")
print(f"\n🎉 所有步骤完成！WordNet Taboo LLM Benchmark 实验流程执行完毕！")
```
