```python
# æ‰§è¡ŒOpenRouterå®éªŒ
print("ğŸš€ æ‰§è¡ŒOpenRouterå®éªŒ...")
print("=" * 60)

# å®éªŒæ‰§è¡Œæ§åˆ¶
RUN_OPENROUTER_EXPERIMENT = False  # è®¾ç½®ä¸ºTrueä»¥è¿è¡ŒçœŸå®å®éªŒ

def create_openrouter_llm_interface(model_name):
    """åˆ›å»ºOpenRouterå…¼å®¹çš„LLMæ¥å£"""
    
    # è·å–APIå¯†é’¥
    openrouter_key = os.getenv('OPENROUTER_API_KEY')
    if not openrouter_key and os.path.exists('api_keys.json'):
        try:
            with open('api_keys.json', 'r') as f:
                keys = json.load(f)
                if 'openrouter_api_key' in keys:
                    openrouter_key = keys['openrouter_api_key']
        except:
            pass
    
    if not openrouter_key:
        raise ValueError("OpenRouter APIå¯†é’¥æœªé…ç½®")
    
    # è·å–OpenRouteræ¨¡å‹åç§°
    openrouter_model = OPENROUTER_CONFIG['models'].get(model_name, model_name)
    
    class OpenRouterLLM:
        def __init__(self, model_name, api_key):
            self.model_name = model_name
            self.api_key = api_key
            self.base_url = OPENROUTER_CONFIG['base_url']
        
        def generate_response(self, prompt, max_tokens=150, temperature=0.7):
            """ç”Ÿæˆå“åº”"""
            headers = {
                'Authorization': f'Bearer {self.api_key}',
                'Content-Type': 'application/json',
                'HTTP-Referer': 'https://github.com/taboo-benchmark',
                'X-Title': 'Taboo LLM Benchmark'
            }
            
            data = {
                'model': self.model_name,
                'messages': [{'role': 'user', 'content': prompt}],
                'max_tokens': max_tokens,
                'temperature': temperature
            }
            
            try:
                response = requests.post(
                    f"{self.base_url}/chat/completions",
                    headers=headers,
                    json=data,
                    timeout=30
                )
                
                if response.status_code == 200:
                    result = response.json()
                    if 'choices' in result and len(result['choices']) > 0:
                        return result['choices'][0]['message']['content']
                    else:
                        return "å“åº”æ ¼å¼é”™è¯¯"
                else:
                    return f"APIé”™è¯¯: {response.status_code}"
                    
            except Exception as e:
                return f"è¯·æ±‚å¤±è´¥: {e}"
    
    return OpenRouterLLM(openrouter_model, openrouter_key)

def run_openrouter_experiment():
    """è¿è¡ŒOpenRouterå®éªŒ"""
    if 'openrouter_exp_config' not in globals():
        print("âŒ OpenRouterå®éªŒé…ç½®æœªå®Œæˆï¼Œè¯·å…ˆè¿è¡Œä¸Šä¸€ä¸ªå•å…ƒæ ¼")
        return None
    
    print("ğŸ¯ å¼€å§‹è¿è¡ŒOpenRouterå®éªŒ...")
    print(f"ğŸ“Š å®éªŒé…ç½®: {len(openrouter_exp_config['models'])} ä¸ªæ¨¡å‹ Ã— {len(openrouter_exp_config['datasets'])} ä¸ªæ•°æ®é›†")
    
    results_summary = []
    total_configs = len(openrouter_exp_config['models']) ** 2 * len(openrouter_exp_config['datasets'])
    current_config = 0
    
    start_time = time.time()
    
    # è¿è¡Œæ‰€æœ‰é…ç½®ç»„åˆ
    for hinter_model in openrouter_exp_config['models']:
        for guesser_model in openrouter_exp_config['models']:
            for dataset in openrouter_exp_config['datasets']:
                current_config += 1
                print(f"\\nğŸ”„ è¿›åº¦: {current_config}/{total_configs} - {hinter_model} â†’ {guesser_model} on {dataset}")
                
                try:
                    config_start_time = time.time()
                    
                    # åˆ›å»ºOpenRouter LLMæ¥å£
                    hinter_llm = create_openrouter_llm_interface(hinter_model)
                    guesser_llm = create_openrouter_llm_interface(guesser_model)
                    
                    # è¿è¡Œå°‘é‡æ¸¸æˆè¿›è¡Œæµ‹è¯•
                    games_played = 0
                    successful_games = 0
                    total_turns = 0
                    
                    # ä»æ•°æ®é›†ä¸­é€‰æ‹©å‡ ä¸ªè¯è¿›è¡Œæµ‹è¯•
                    dataset_items = experiment.datasets[dataset]
                    test_items = dataset_items.sample(min(openrouter_exp_config['games_per_config'], len(dataset_items)))
                    
                    for item in test_items:
                        games_played += 1
                        
                        # æ¨¡æ‹Ÿæ¸¸æˆç»“æœï¼ˆç®€åŒ–ç‰ˆï¼‰
                        target_word = item.get('target', 'test')
                        taboo_words = item.get('taboo', [])
                        
                        # ç®€å•çš„æµ‹è¯•ï¼šè®©hinteræè¿°è¯æ±‡
                        hinter_prompt = f"è¯·ç”¨ç®€çŸ­çš„å¥å­æè¿°'{target_word}'ï¼Œä½†ä¸èƒ½ä½¿ç”¨è¿™äº›è¯: {', '.join(taboo_words[:3])}"
                        hinter_response = hinter_llm.generate_response(hinter_prompt, max_tokens=50)
                        
                        # è®©guesserçŒœæµ‹
                        guesser_prompt = f"æ ¹æ®è¿™ä¸ªæè¿°ï¼ŒçŒœæµ‹ç›®æ ‡è¯æ±‡æ˜¯ä»€ä¹ˆï¼š{hinter_response}"
                        guesser_response = guesser_llm.generate_response(guesser_prompt, max_tokens=20)
                        
                        # ç®€å•åˆ¤æ–­æ˜¯å¦æˆåŠŸï¼ˆåŒ…å«ç›®æ ‡è¯ï¼‰
                        if target_word.lower() in guesser_response.lower():
                            successful_games += 1
                            total_turns += 1
                        else:
                            total_turns += 2
                    
                    config_end_time = time.time()
                    config_duration = config_end_time - config_start_time
                    
                    result_summary = {
                        'hinter_model': hinter_model,
                        'guesser_model': guesser_model,
                        'dataset': dataset,
                        'success_rate': successful_games / games_played if games_played > 0 else 0,
                        'total_games': games_played,
                        'avg_turns': total_turns / games_played if games_played > 0 else 0,
                        'duration_seconds': config_duration
                    }
                    results_summary.append(result_summary)
                    
                    print(f"   âœ… å®Œæˆ - æˆåŠŸç‡: {result_summary['success_rate']:.1%}, ç”¨æ—¶: {config_duration:.1f}ç§’")
                    
                except Exception as e:
                    print(f"   âŒ å¤±è´¥: {e}")
                    continue
    
    end_time = time.time()
    total_duration = end_time - start_time
    
    if results_summary:
        # åˆ›å»ºç»“æœæ•°æ®æ¡†
        results_df = pd.DataFrame(results_summary)
        
        # ä¿å­˜ç»“æœ
        timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
        results_file = f"openrouter_experiment_results_{timestamp}.csv"
        results_df.to_csv(results_file, index=False)
        
        print(f"\\nğŸ“Š OpenRouterå®éªŒå®Œæˆæ‘˜è¦:")
        print(f"   - æˆåŠŸé…ç½®æ•°: {len(results_df)}")
        print(f"   - æ€»ç”¨æ—¶: {total_duration:.1f} ç§’ ({total_duration/60:.1f} åˆ†é’Ÿ)")
        print(f"   - å¹³å‡æ¯é…ç½®ç”¨æ—¶: {total_duration/len(results_df):.1f} ç§’")
        print(f"   - ç»“æœæ–‡ä»¶: {results_file}")
        
        return results_df
    else:
        print("\\nâš ï¸ æ²¡æœ‰æˆåŠŸå®Œæˆä»»ä½•å®éªŒé…ç½®")
        return None

# å®éªŒæ‰§è¡Œ
if RUN_OPENROUTER_EXPERIMENT:
    if 'openrouter_exp_config' in globals():
        print("âš ï¸ å³å°†å¼€å§‹OpenRouterçœŸå®å®éªŒï¼Œè¿™å°†äº§ç”ŸAPIè°ƒç”¨è´¹ç”¨")
        print("ğŸ’° é¢„è®¡æˆæœ¬:")
        
        total_games = len(openrouter_exp_config['models']) ** 2 * len(openrouter_exp_config['datasets']) * openrouter_exp_config['games_per_config']
        api_calls = total_games * 4  # æ¯æ¸¸æˆçº¦4æ¬¡APIè°ƒç”¨ï¼ˆç®€åŒ–ç‰ˆï¼‰
        
        print(f"   - é¢„è®¡APIè°ƒç”¨æ•°: {api_calls}")
        print(f"   - OpenRouterç»Ÿä¸€è®¡è´¹ï¼Œè´¹ç”¨é€æ˜")
        
        print("\\nğŸš€ å¼€å§‹OpenRouterå®éªŒ...")
        openrouter_experiment_results = run_openrouter_experiment()
        
        if openrouter_experiment_results is not None:
            print("\\nâœ… OpenRouterå®éªŒæ‰§è¡Œå®Œæˆï¼")
        else:
            print("\\nâŒ OpenRouterå®éªŒæ‰§è¡Œå¤±è´¥")
    else:
        print("âŒ è¯·å…ˆè¿è¡ŒOpenRouterå®éªŒé…ç½®")
        
else:
    print("ğŸ’¡ è¦è¿è¡ŒOpenRouterçœŸå®å®éªŒï¼Œè¯·æ‰§è¡Œä»¥ä¸‹æ­¥éª¤:")
    print("1. ç¡®ä¿å·²å®Œæˆæ­¥éª¤1å’Œæ­¥éª¤2")
    print("2. è®¾ç½® RUN_OPENROUTER_EXPERIMENT = True")
    print("3. é‡æ–°è¿è¡Œæ­¤å•å…ƒæ ¼")
    print("\\nâš ï¸ æ³¨æ„ï¼šè¿è¡ŒçœŸå®å®éªŒå°†äº§ç”ŸOpenRouter APIè°ƒç”¨è´¹ç”¨")
    print("ğŸ’¡ å»ºè®®å…ˆè¿›è¡Œå°è§„æ¨¡æµ‹è¯•ï¼ˆå‡å°‘games_per_configçš„å€¼ï¼‰")

print(f"\\nğŸ” æ­¥éª¤3.2å®Œæˆï¼šOpenRouterå®éªŒæ‰§è¡Œæ§åˆ¶å°±ç»ªï¼")
```

```python
# æ­¥éª¤4: åˆ›å»ºæ¼”ç¤ºæ•°æ®ç”¨äºåˆ†æå±•ç¤º
print("ğŸ“Š æ­¥éª¤4ï¼šOpenRouterå®éªŒç»“æœåˆ†æ...")
print("=" * 60)

# æ£€æŸ¥æ˜¯å¦æœ‰çœŸå®å®éªŒç»“æœ
if 'openrouter_experiment_results' in globals() and openrouter_experiment_results is not None:
    print("âœ… å‘ç°OpenRouterçœŸå®å®éªŒç»“æœï¼Œå°†ä½¿ç”¨çœŸå®æ•°æ®è¿›è¡Œåˆ†æ")
    df = openrouter_experiment_results.copy()
    data_source = "OpenRouterçœŸå®å®éªŒæ•°æ®"
else:
    print("ğŸ“ æœªå‘ç°çœŸå®å®éªŒç»“æœï¼Œåˆ›å»ºåŸºäºOpenRouteræ¨¡å‹çš„æ¼”ç¤ºæ•°æ®...")
    
    # åˆ›å»ºåŸºäºOpenRouteræ¨¡å‹çš„æ¼”ç¤ºæ•°æ®
    np.random.seed(42)
    demo_data = []
    
    # è·å–WordNetæ•°æ®é›†çš„è¯æ±‡
    wordnet_words = []
    if 'experiment' in globals() and 'wordnet' in experiment.datasets:
        wordnet_dataset = experiment.datasets['wordnet']
        wordnet_words = [item.get('target', f'word_{i}') for i, item in enumerate(wordnet_dataset[:50])]
    else:
        wordnet_words = [f"wordnet_word_{i}" for i in range(50)]
    
    # OpenRouteræ”¯æŒçš„æ¨¡å‹
    openrouter_models = list(OPENROUTER_CONFIG['models'].keys())[:6]  # ä½¿ç”¨å‰6ä¸ªæ¨¡å‹
    datasets = ["wordnet", "polysemous_wordnet", "general"]
    
    print(f"ğŸ¯ ç”ŸæˆåŸºäº {len(wordnet_words)} ä¸ªWordNetè¯æ±‡å’Œ {len(openrouter_models)} ä¸ªOpenRouteræ¨¡å‹çš„æ¼”ç¤ºæ•°æ®...")
    
    for hinter in openrouter_models[:4]:  # ä½¿ç”¨å‰4ä¸ªæ¨¡å‹
        for guesser in openrouter_models[:4]:
            for dataset in datasets:
                for i in range(8):  # æ¯ä¸ªé…ç½®8ä¸ªæ¸¸æˆ
                    # ä¸ºä¸åŒæ¨¡å‹è®¾ç½®ä¸åŒçš„æˆåŠŸç‡
                    if 'gpt-4' in hinter or 'claude' in hinter:
                        base_success_prob = 0.80  # é«˜çº§æ¨¡å‹æˆåŠŸç‡è¾ƒé«˜
                    elif 'gpt-3.5' in hinter or 'gemini' in hinter:
                        base_success_prob = 0.70  # ä¸­çº§æ¨¡å‹
                    else:
                        base_success_prob = 0.60  # å…¶ä»–æ¨¡å‹
                    
                    # ä¸ºä¸åŒæ•°æ®é›†è°ƒæ•´æˆåŠŸç‡
                    if dataset == "wordnet":
                        success_prob = base_success_prob * 1.1  # WordNetæ•°æ®é›†ç•¥å®¹æ˜“
                    elif dataset == "polysemous_wordnet":
                        success_prob = base_success_prob * 0.85  # å¤šä¹‰è¯æ›´å›°éš¾
                    else:
                        success_prob = base_success_prob
                    
                    success_prob = min(success_prob, 0.95)  # é™åˆ¶æœ€é«˜æˆåŠŸç‡
                    
                    success = np.random.choice([True, False], p=[success_prob, 1-success_prob])
                    turns = np.random.randint(1, 4) if success else np.random.randint(2, 5)
                    
                    # éšæœºé€‰æ‹©ä¸€ä¸ªWordNetè¯æ±‡
                    target_word = np.random.choice(wordnet_words)
                    
                    demo_data.append({
                        'hinter_model': hinter,
                        'guesser_model': guesser,
                        'dataset': dataset,
                        'target_word': target_word,
                        'success': success,
                        'num_turns': turns,
                        'game_duration': np.random.uniform(30, 120),  # 30-120ç§’
                        'openrouter_model_hinter': OPENROUTER_CONFIG['models'].get(hinter, hinter),
                        'openrouter_model_guesser': OPENROUTER_CONFIG['models'].get(guesser, guesser)
                    })
    
    df = pd.DataFrame(demo_data)
    data_source = "OpenRouteræ¼”ç¤ºæ•°æ®"
    print(f"âœ… åˆ›å»ºäº† {len(df)} æ¡åŸºäºOpenRouteræ¨¡å‹çš„æ¼”ç¤ºæ¸¸æˆè®°å½•")

print(f"\\nğŸ“Š æ•°æ®åˆ†æä½¿ç”¨: {data_source}")
print(f"ğŸ“‹ æ•°æ®æ¦‚è§ˆ:")
print(f"   - æ€»è®°å½•æ•°: {len(df)}")
print(f"   - æ•°æ®é›†ç±»å‹: {df['dataset'].unique()}")
print(f"   - æ¨¡å‹æ•°é‡: {len(df['hinter_model'].unique())}")

# åŸºç¡€ç»Ÿè®¡
if 'success' in df.columns:
    overall_success_rate = df['success'].mean()
    print(f"   - æ€»ä½“æˆåŠŸç‡: {overall_success_rate:.1%}")
elif 'success_rate' in df.columns:
    overall_success_rate = df['success_rate'].mean()
    print(f"   - å¹³å‡æˆåŠŸç‡: {overall_success_rate:.1%}")

# æ˜¾ç¤ºæ•°æ®é›†åˆ†å¸ƒ
print(f"\\nğŸ“Š æ•°æ®åˆ†å¸ƒ:")
dataset_counts = df['dataset'].value_counts()
for dataset, count in dataset_counts.items():
    if 'success' in df.columns:
        success_rate = df[df['dataset'] == dataset]['success'].mean()
        print(f"   - {dataset}: {count} åœºæ¸¸æˆ, æˆåŠŸç‡: {success_rate:.1%}")
    else:
        print(f"   - {dataset}: {count} åœºæ¸¸æˆ")

# æ˜¾ç¤ºOpenRouteræ¨¡å‹åˆ†å¸ƒ
print(f"\\nğŸ¤– OpenRouteræ¨¡å‹ä½¿ç”¨æƒ…å†µ:")
model_counts = df['hinter_model'].value_counts()
for model, count in model_counts.items():
    openrouter_model = OPENROUTER_CONFIG['models'].get(model, model)
    provider = openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    print(f"   - {model} ({provider}): {count} æ¬¡")

# æ˜¾ç¤ºç¤ºä¾‹è®°å½•
print(f"\\nğŸ“‹ ç¤ºä¾‹è®°å½• (å‰5æ¡):")
if 'success' in df.columns:
    display_cols = ['hinter_model', 'guesser_model', 'dataset', 'success', 'num_turns']
else:
    display_cols = ['hinter_model', 'guesser_model', 'dataset', 'success_rate', 'total_games']

print(df[display_cols].head().to_string(index=False))

print(f"\\nğŸ” æ­¥éª¤4.1å®Œæˆï¼šOpenRouteræ•°æ®å‡†å¤‡å®Œæ¯•ï¼")
```

```python
# OpenRouteræ¨¡å‹æ€§èƒ½åˆ†æ
print("ğŸ“ˆ OpenRouteræ¨¡å‹æ€§èƒ½åˆ†æ...")
print("=" * 60)

# æŒ‰æä¾›å•†åˆ†ææ¨¡å‹æ€§èƒ½
def analyze_openrouter_performance(df):
    """åˆ†æOpenRouteræ¨¡å‹æ€§èƒ½ï¼ŒæŒ‰æä¾›å•†åˆ†ç»„"""
    
    # æ·»åŠ æä¾›å•†ä¿¡æ¯
    df_analysis = df.copy()
    
    def get_provider(model_name):
        openrouter_model = OPENROUTER_CONFIG['models'].get(model_name, model_name)
        return openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    
    df_analysis['hinter_provider'] = df_analysis['hinter_model'].apply(get_provider)
    df_analysis['guesser_provider'] = df_analysis['guesser_model'].apply(get_provider)
    
    # æ ¹æ®æ•°æ®ç»“æ„è°ƒæ•´åˆ†æé€»è¾‘
    if 'success' in df_analysis.columns:
        # ä¸ªåˆ«æ¸¸æˆæ•°æ®
        hinter_performance = df_analysis.groupby('hinter_model')['success'].agg(['mean', 'count']).round(3)
        hinter_performance.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
        
        guesser_performance = df_analysis.groupby('guesser_model')['success'].agg(['mean', 'count']).round(3)
        guesser_performance.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
        
        # æŒ‰æä¾›å•†åˆ†æ
        provider_hinter_perf = df_analysis.groupby('hinter_provider')['success'].agg(['mean', 'count']).round(3)
        provider_hinter_perf.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
        
        provider_guesser_perf = df_analysis.groupby('guesser_provider')['success'].agg(['mean', 'count']).round(3)
        provider_guesser_perf.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
        
    else:
        # èšåˆå®éªŒç»“æœæ•°æ®
        if 'success_rate' in df_analysis.columns:
            hinter_performance = df_analysis.groupby('hinter_model')['success_rate'].agg(['mean', 'count']).round(3)
            hinter_performance.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']
            
            guesser_performance = df_analysis.groupby('guesser_model')['success_rate'].agg(['mean', 'count']).round(3)
            guesser_performance.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']
            
            provider_hinter_perf = df_analysis.groupby('hinter_provider')['success_rate'].agg(['mean', 'count']).round(3)
            provider_hinter_perf.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']
            
            provider_guesser_perf = df_analysis.groupby('guesser_provider')['success_rate'].agg(['mean', 'count']).round(3)
            provider_guesser_perf.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']
    
    return hinter_performance, guesser_performance, provider_hinter_perf, provider_guesser_perf

# æ‰§è¡Œæ€§èƒ½åˆ†æ
hinter_perf, guesser_perf, provider_hinter_perf, provider_guesser_perf = analyze_openrouter_performance(df)

print("ğŸ¯ æç¤ºè€…(Hinter)æ¨¡å‹æ€§èƒ½:")
for model, row in hinter_perf.sort_values(hinter_perf.columns[0], ascending=False).iterrows():
    openrouter_model = OPENROUTER_CONFIG['models'].get(model, model)
    provider = openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    print(f"   {model} ({provider}): {row.iloc[0]:.1%} ({row.iloc[1]} æ¬¡)")

print("\\nğŸ” çŒœæµ‹è€…(Guesser)æ¨¡å‹æ€§èƒ½:")
for model, row in guesser_perf.sort_values(guesser_perf.columns[0], ascending=False).iterrows():
    openrouter_model = OPENROUTER_CONFIG['models'].get(model, model)
    provider = openrouter_model.split('/')[0] if '/' in openrouter_model else 'unknown'
    print(f"   {model} ({provider}): {row.iloc[0]:.1%} ({row.iloc[1]} æ¬¡)")

print("\\nğŸ¢ æŒ‰æä¾›å•†åˆ†æ - æç¤ºè€…æ€§èƒ½:")
for provider, row in provider_hinter_perf.sort_values(provider_hinter_perf.columns[0], ascending=False).iterrows():
    print(f"   {provider}: {row.iloc[0]:.1%} ({row.iloc[1]} æ¬¡)")

print("\\nğŸ¢ æŒ‰æä¾›å•†åˆ†æ - çŒœæµ‹è€…æ€§èƒ½:")
for provider, row in provider_guesser_perf.sort_values(provider_guesser_perf.columns[0], ascending=False).iterrows():
    print(f"   {provider}: {row.iloc[0]:.1%} ({row.iloc[1]} æ¬¡)")

# æ•°æ®é›†éš¾åº¦åˆ†æ
if 'success' in df.columns:
    dataset_performance = df.groupby('dataset')['success'].agg(['mean', 'count']).round(3)
    dataset_performance.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
elif 'success_rate' in df.columns:
    dataset_performance = df.groupby('dataset')['success_rate'].agg(['mean', 'count']).round(3)
    dataset_performance.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']

print("\\nğŸ“Š æ•°æ®é›†éš¾åº¦åˆ†æ:")
for dataset, row in dataset_performance.sort_values(dataset_performance.columns[0], ascending=False).iterrows():
    print(f"   {dataset}: {row.iloc[0]:.1%} ({row.iloc[1]} æ¬¡)")

# æ¨¡å‹ç»„åˆåˆ†æ
if 'success' in df.columns:
    print("\\nğŸ¤ æœ€ä½³OpenRouteræ¨¡å‹ç»„åˆ (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['æˆåŠŸç‡', 'æ¸¸æˆæ•°']
    top_combinations = combination_perf.sort_values('æˆåŠŸç‡', ascending=False).head()
    for (hinter, guesser), row in top_combinations.iterrows():
        hinter_provider = OPENROUTER_CONFIG['models'].get(hinter, hinter).split('/')[0]
        guesser_provider = OPENROUTER_CONFIG['models'].get(guesser, guesser).split('/')[0]
        print(f"   {hinter} ({hinter_provider}) â†’ {guesser} ({guesser_provider}): {row['æˆåŠŸç‡']:.1%}")
elif 'success_rate' in df.columns:
    print("\\nğŸ¤ æœ€ä½³OpenRouteræ¨¡å‹ç»„åˆ (Top 5):")
    combination_perf = df.groupby(['hinter_model', 'guesser_model'])['success_rate'].agg(['mean', 'count']).round(3)
    combination_perf.columns = ['å¹³å‡æˆåŠŸç‡', 'é…ç½®æ•°']
    top_combinations = combination_perf.sort_values('å¹³å‡æˆåŠŸç‡', ascending=False).head()
    for (hinter, guesser), row in top_combinations.iterrows():
        hinter_provider = OPENROUTER_CONFIG['models'].get(hinter, hinter).split('/')[0]
        guesser_provider = OPENROUTER_CONFIG['models'].get(guesser, guesser).split('/')[0]
        print(f"   {hinter} ({hinter_provider}) â†’ {guesser} ({guesser_provider}): {row['å¹³å‡æˆåŠŸç‡']:.1%}")

print(f"\\nğŸ” æ­¥éª¤4.2å®Œæˆï¼šOpenRouteræ€§èƒ½åˆ†æå®Œæ¯•ï¼")
```

```python
# è¯¦ç»†ç»“æœæŠ¥å‘Šå’Œå¯¼å‡º
print("ğŸ“„ è¯¦ç»†ç»“æœæŠ¥å‘Šå’Œå¯¼å‡º...")
print("=" * 60)

# ç”Ÿæˆè¯¦ç»†æŠ¥å‘Š
def generate_detailed_report(df):
    """ç”Ÿæˆè¯¦ç»†çš„å®éªŒæŠ¥å‘Š"""
    
    report = {
        'experiment_summary': {},
        'model_analysis': {},
        'dataset_analysis': {},
        'recommendations': []
    }
    
    # å®éªŒæ‘˜è¦
    report['experiment_summary'] = {
        'total_records': len(df),
        'unique_models': len(df['hinter_model'].unique()),
        'unique_datasets': len(df['dataset'].unique()),
        'data_source': data_source
    }
    
    if 'success' in df.columns:
        report['experiment_summary']['overall_success_rate'] = f"{df['success'].mean():.1%}"
        report['experiment_summary']['total_games'] = len(df)
    elif 'success_rate' in df.columns:
        report['experiment_summary']['avg_success_rate'] = f"{df['success_rate'].mean():.1%}"
        if 'total_games' in df.columns:
            report['experiment_summary']['total_games'] = df['total_games'].sum()
    
    # æ¨¡å‹åˆ†æ
    if 'success' in df.columns:
        hinter_perf = df.groupby('hinter_model')['success'].mean().sort_values(ascending=False)
        guesser_perf = df.groupby('guesser_model')['success'].mean().sort_values(ascending=False)
    else:
        hinter_perf = df.groupby('hinter_model')['success_rate'].mean().sort_values(ascending=False)
        guesser_perf = df.groupby('guesser_model')['success_rate'].mean().sort_values(ascending=False)
    
    report['model_analysis'] = {
        'best_hinter': hinter_perf.index[0],
        'best_hinter_score': f"{hinter_perf.iloc[0]:.1%}",
        'best_guesser': guesser_perf.index[0],
        'best_guesser_score': f"{guesser_perf.iloc[0]:.1%}",
        'hinter_ranking': hinter_perf.to_dict(),
        'guesser_ranking': guesser_perf.to_dict()
    }
    
    # æ•°æ®é›†åˆ†æ
    if 'success' in df.columns:
        dataset_perf = df.groupby('dataset')['success'].mean().sort_values(ascending=False)
    else:
        dataset_perf = df.groupby('dataset')['success_rate'].mean().sort_values(ascending=False)
    
    report['dataset_analysis'] = {
        'easiest_dataset': dataset_perf.index[0],
        'easiest_score': f"{dataset_perf.iloc[0]:.1%}",
        'hardest_dataset': dataset_perf.index[-1],
        'hardest_score': f"{dataset_perf.iloc[-1]:.1%}",
        'dataset_ranking': dataset_perf.to_dict()
    }
    
    # ç”Ÿæˆå»ºè®®
    recommendations = []
    
    if report['model_analysis']['best_hinter_score'] > "75%":
        recommendations.append(f"æ¨èä½¿ç”¨ {report['model_analysis']['best_hinter']} ä½œä¸ºæç¤ºè€…æ¨¡å‹")
    
    if report['model_analysis']['best_guesser_score'] > "75%":
        recommendations.append(f"æ¨èä½¿ç”¨ {report['model_analysis']['best_guesser']} ä½œä¸ºçŒœæµ‹è€…æ¨¡å‹")
    
    if report['dataset_analysis']['easiest_score'] > "80%":
        recommendations.append(f"{report['dataset_analysis']['easiest_dataset']} æ•°æ®é›†é€‚åˆæ¨¡å‹è®­ç»ƒå’Œè¯„ä¼°")
    
    if report['dataset_analysis']['hardest_score'] < "60%":
        recommendations.append(f"{report['dataset_analysis']['hardest_dataset']} æ•°æ®é›†å…·æœ‰æŒ‘æˆ˜æ€§ï¼Œé€‚åˆé«˜çº§æ¨¡å‹æµ‹è¯•")
    
    report['recommendations'] = recommendations
    
    return report

# ç”ŸæˆæŠ¥å‘Š
detailed_report = generate_detailed_report(df)

print("ğŸ“Š å®éªŒæ€»ç»“æŠ¥å‘Š")
print(f"   - æ•°æ®æ¥æº: {detailed_report['experiment_summary']['data_source']}")
print(f"   - æ€»è®°å½•æ•°: {detailed_report['experiment_summary']['total_records']}")
print(f"   - æµ‹è¯•æ¨¡å‹æ•°: {detailed_report['experiment_summary']['unique_models']}")
print(f"   - æµ‹è¯•æ•°æ®é›†æ•°: {detailed_report['experiment_summary']['unique_datasets']}")

if 'overall_success_rate' in detailed_report['experiment_summary']:
    print(f"   - æ€»ä½“æˆåŠŸç‡: {detailed_report['experiment_summary']['overall_success_rate']}")
elif 'avg_success_rate' in detailed_report['experiment_summary']:
    print(f"   - å¹³å‡æˆåŠŸç‡: {detailed_report['experiment_summary']['avg_success_rate']}")

print(f"\nğŸ† æœ€ä½³æ¨¡å‹è¡¨ç°:")
print(f"   - æœ€ä½³æç¤ºè€…: {detailed_report['model_analysis']['best_hinter']} ({detailed_report['model_analysis']['best_hinter_score']})")
print(f"   - æœ€ä½³çŒœæµ‹è€…: {detailed_report['model_analysis']['best_guesser']} ({detailed_report['model_analysis']['best_guesser_score']})")

print(f"\nğŸ“Š æ•°æ®é›†éš¾åº¦:")
print(f"   - æœ€å®¹æ˜“: {detailed_report['dataset_analysis']['easiest_dataset']} ({detailed_report['dataset_analysis']['easiest_score']})")
print(f"   - æœ€å›°éš¾: {detailed_report['dataset_analysis']['hardest_dataset']} ({detailed_report['dataset_analysis']['hardest_score']})")

print(f"\nğŸ’¡ å»ºè®®:")
for i, recommendation in enumerate(detailed_report['recommendations'], 1):
    print(f"   {i}. {recommendation}")

# å¯¼å‡ºç»“æœ
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')

# å¯¼å‡ºCSV
csv_filename = f"taboo_benchmark_results_{timestamp}.csv"
df.to_csv(csv_filename, index=False, encoding='utf-8')
print(f"\nğŸ“ ç»“æœå·²å¯¼å‡º:")
print(f"   - CSVæ–‡ä»¶: {csv_filename}")

# å¯¼å‡ºExcelï¼ˆå¦‚æœæ•°æ®ä¸å¤ªå¤§ï¼‰
if len(df) < 10000:
    try:
        excel_filename = f"taboo_benchmark_results_{timestamp}.xlsx"
        with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
            df.to_excel(writer, sheet_name='åŸå§‹æ•°æ®', index=False)
            
            # æ·»åŠ æ±‡æ€»å·¥ä½œè¡¨
            if 'success' in df.columns:
                summary_data = {
                    'æ¨¡å‹': list(hinter_perf.index),
                    'æç¤ºè€…æˆåŠŸç‡': [f"{x:.1%}" for x in hinter_perf.values],
                    'çŒœæµ‹è€…æˆåŠŸç‡': [f"{guesser_perf.loc[model]:.1%}" for model in hinter_perf.index]
                }
            else:
                summary_data = {
                    'æ¨¡å‹': list(hinter_perf.index),
                    'æç¤ºè€…å¹³å‡æˆåŠŸç‡': [f"{x:.1%}" for x in hinter_perf.values],
                    'çŒœæµ‹è€…å¹³å‡æˆåŠŸç‡': [f"{guesser_perf.loc[model]:.1%}" for model in hinter_perf.index]
                }
            
            summary_df = pd.DataFrame(summary_data)
            summary_df.to_excel(writer, sheet_name='æ¨¡å‹æ€§èƒ½æ±‡æ€»', index=False)
        
        print(f"   - Excelæ–‡ä»¶: {excel_filename}")
    except Exception as e:
        print(f"   - Excelå¯¼å‡ºå¤±è´¥: {e}")

# ä¿å­˜è¯¦ç»†æŠ¥å‘Š
report_filename = f"taboo_benchmark_report_{timestamp}.json"
with open(report_filename, 'w', encoding='utf-8') as f:
    import json
    json.dump(detailed_report, f, ensure_ascii=False, indent=2)

print(f"   - è¯¦ç»†æŠ¥å‘Š: {report_filename}")

print(f"\nğŸ” æ­¥éª¤4.4å®Œæˆï¼šæŠ¥å‘Šç”Ÿæˆå’Œå¯¼å‡ºå®Œæ¯•ï¼")
print(f"\nğŸ‰ æ‰€æœ‰æ­¥éª¤å®Œæˆï¼WordNet Taboo LLM Benchmark å®éªŒæµç¨‹æ‰§è¡Œå®Œæ¯•ï¼")
```
